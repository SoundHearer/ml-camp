### 2.某评测网站对市场上某品牌充电宝随机抽取30款测得其使用寿命数据，请问哪些可视化方案适宜呈现该结果？这些方案透露出了哪些含义

### 3.	一组数据的分布特征可以从哪几个方面进行描述？

1.分布的集中趋势，反映各数据向其中心值靠拢或聚集的程度。
2.分布的离散程度，反映各数据远离其中心值的趋势。
3.分布的形状，反映数据分布的偏态和峰态。

### 4.（接上题）如何从可视化的角度解读一组数据的分布特征？(代码演示有额外加分)

```python
# 如下所示，箱型图显示了数据分不多集中与离散趋势，kde曲线展示了分布的形状
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
x = np.random.normal(size=100)
plt.subplot(2,1,1)
sns.distplot(x,kde=True, rug=True, bins=30);
plt.subplot(2,1,2)
sns.boxplot(x)
```

### 5.简要谈谈您对seaborn库的理解，以及有哪些较高级的用法。(代码演示有额外加分)

Seaborn在matplotlib的基础上进行了更高级的API封装使数据可视化更简单和便捷，还集成了统计和展示数据分布特征的功能。
比如seaborn可以对数据集做多项式拟合，还可以在拟合时通过传入传入robust=True排除一些噪点干扰。还有FacetGrid可以绘制最多三个维度：row，col和hue。
PairGrid可以快速直观的汇总并展示出数据集中，任意两个维度间的关系。

```python
anscombe = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80});


sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           ci=None, scatter_kws={"s": 80},robust=True);


iris = sns.load_dataset("iris")
g = sns.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot, cmap="Blues_d")
g.map_diag(plt.hist)
```

### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
iris = sns.load_dataset("iris")
iris['sepal_size']=iris['sepal_length']*iris['sepal_width']
iris['petal_size']=iris['petal_length']*iris['petal_width']
f1=sns.jointplot(x='sepal_size',y='petal_size',data=iris)
```

```python
sepal=iris['sepal_size']
petal=iris['petal_size']
species=iris['species']
species=species.append(species)
size=sepal.append(petal)
c1=np.zeros(len(sepal))
c2=np.ones(len(petal))
c= np.append(c1,c2)
df=pd.DataFrame({'size':size, 'class':c, 'species':species})
df['class'] = df['class'].replace({0.0:'sepal', 1.0:'petal'})
sns.boxplot(y='size',x='species', hue="class",data=df)
```

```python
sns.factorplot(x="sepal_size", y="petal_size", col="species", data=iris, kind="swarm")
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
titanic = sns.load_dataset("titanic")
sur = titanic.groupby('class')['survived'].sum()
total = titanic.groupby('class')['survived'].count()
sns.set_context({"figure.figsize": (6, 5)})
topbar = sns.barplot(x = total.index, y = total, color = "red",label='unsuvived')
bottombar = sns.barplot(x = sur.index, y = sur, color = "#0000A3",label='survived')
plt.legend(loc='upper left',fontsize =12)
```

```python
titanic = sns.load_dataset("titanic")
sur = titanic.groupby('sex')['survived'].sum()
total = titanic.groupby('sex')['survived'].count()
sns.set_context({"figure.figsize": (5, 5)})
topbar = sns.barplot(x = total.index, y = total, color = "red",label='unsuvived')
bottombar = sns.barplot(x = sur.index, y = sur, color = "#0000A3",label='survived')
plt.legend(loc='upper left',fontsize =12)
```
