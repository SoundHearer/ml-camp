### 4.（接上题）如何从可视化的角度解读一组数据的分布特征？(代码演示有额外加分)

通过可视化可以看出样本分布的形状, 均值, 以及波动性

```python
%matplotlib inline
%config InlineBackend.figure_format = "png"
import seaborn as sns
import numpy as np
import matplotlib as mp
import matplotlib.pylab as plt
mp.rcParams['font.family'] = 'Liberation Mono'

x1 = np.random.normal(2,0.2,size=300)
x2 = np.random.normal(size=300)

sns.distplot(x1,kde=False,bins=50, label="x1")
sns.distplot(x2,kde=False,bins=50, label="x2")
plt.legend();
```

### 1.航班乘客变化分析

- 载入相应绘图库及航班乘客数据（flights）

```python
# your code
df = sns.load_dataset("flights")
df.head()
```

- 分析年度乘客总量变化情况（折线图）

```python
# your code
data=df.groupby("year").sum()
data["year"] = data.index
sns.pointplot(x = "year",y = "passengers",estimator = np.sum, data=data)
```

- 分析乘客在一年中各月份的分布（柱状图）

```python
# your code
_, ax = plt.subplots(figsize=(12,4))
sns.barplot(x="month",y="passengers",data=df, ax=ax, estimator = np.sum)
```

### 2.鸢尾花花型尺寸分析

* 载入iris数据集


```python
df = sns.load_dataset("iris")
df.head()
```

* 萼片（sepal）和花瓣（petal）的大小关系（散点图）


```python
df["sepal"] = (df["sepal_length"].values**2+df["sepal_width"].values**2)**0.5
df["petal"] = (df["petal_length"].values**2+df["petal_width"].values**2)**0.5
sns.jointplot(x = "sepal", y ="petal",data=df)
```

- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）


```python
sns.boxplot(x="species", y="petal",data=df)
```

* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
sns.stripplot(x="species", y="petal", data=df, jitter=True)
```

### 3.餐厅小费情况分析
* 载入tips数据集


```python
# your code
df = sns.load_dataset("tips")
df.head()
```

* 小费和总消费之间的关系（散点图）

* 男性顾客和女性顾客，谁更慷慨（分类箱式图）

```python
sns.boxplot(x="smoker",y="tip", data= df)
```

```python
df["is_weekend"] = df.day.map(lambda x: "weekend" if x in "Sun Sat" else "work day")
sns.boxplot(x="is_weekend",y="tip", data=df)
```

* 午饭和晚饭，哪一顿顾客更愿意给小费（分类箱式图）


```python
sns.boxplot(x="time",y="tip", data=df)
```

```python
sns.boxplot(x="size",y="tip", data=df)
```

```python
sns.barplot(x="smoker",y="tip",hue="sex", data=df)
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
data = df.groupby("pclass").agg({"survived":lambda x:x.sum()/x.shape[0]})
data["pclass"] = data.index
data["dead"] = 1 - data["survived"]
_, ax = plt.subplots()
sns.barplot(x = "pclass",y = "survived",data=data,color='b',ax=ax, label="survived")
sns.barplot(x = "pclass", y="dead",data=data ,color="g", ax=ax, label="dead", bottom = data["survived"])
plt.legend()
ax.set_ylabel("")
```

* 不同性别的幸存比例（堆积柱状图）

```python
data = df.groupby("sex").agg({"survived":lambda x:x.sum()/x.shape[0]})
data["sex"] = data.index
data["dead"] = 1 - data["survived"]
_, ax = plt.subplots()
sns.barplot(x = "sex",y = "survived",data=data,color='b',ax=ax, label="survived")
sns.barplot(x = "sex", y="dead",data=data ,color="g", ax=ax, label="dead", bottom = data["survived"])
plt.legend()
ax.set_ylabel("")
```

* 幸存和遇难乘客的票价分布（分类箱式图）

```python
_, ax = plt.subplots(figsize=(3,6))
sns.boxplot(x="survived", y = "fare", data=df, ax=ax)
```

* 幸存和遇难乘客的年龄分布（分类箱式图）


```python
sns.boxplot(x="survived",y="age", data=df)
```

* 不同上船港口的乘客仓位等级分布（分组柱状图）

```python
df["count"] = np.ones(df.shape[0])
sns.barplot(x="pclass", y="count", hue="embark_town", data= df, estimator=np.sum)
```

* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）

```python
sns.boxplot(x="survived",y="sibsp", data=df)
```

* 幸存和遇难乘客父母子女的数量分布（分类箱式图）

```python
sns.boxplot(x="survived",y="parch", data=df)
```

* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
# your code
sns.barplot(x="alone", y="count", hue="survived", data= df, estimator=np.sum)
```
