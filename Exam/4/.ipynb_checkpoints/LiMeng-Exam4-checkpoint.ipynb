### 1.谈谈您对数据可视，以及其对数据分析，机器学习的意义的理解

***数据可视就是更为直观的展现数据，若仅仅看数据本身，无法直接获取各个指标的间的关系。通过数据可视化后，将数据转化为人可以直接读懂的图表，借助色彩的差异、图形的走势变化从而直观的展现指标间的关系。对于机器学习来说，数据预处理的结果对建模计算后的最终效果有直接影响，因此，特征的处理非常重要。而人无法在大量数据中获取到某特征下的数值是否分布合理，是否存在一些非正常的值，也无法直接看出特征间的关联。借助可视化则可以解决此问题。数据可视化对数据分析以及机器学习是不可或缺的一步。

### 5.简要谈谈您对seaborn库的理解，以及有哪些较高级的用法。(代码演示有额外加分)

***seaborn库是基于matplotlib开发，对matplotlib众多绘图接口进行了封装，使得繁杂的绘图步骤变得更加简单。但是，灵活性也变低了。seaborn较高级的用法有hue参数进行分组，自动添加kde曲线进行拟合，pairplot可以多维度绘图，lmplot更是可以指定logistic=True使用逻辑回归拟合曲线，而无需手动计算任何参数。案例：

```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns

x = np.random.normal(size=100) # 高斯分布
sns.distplot(x, kde=True) #kde 曲线拟合
```

```python
iris = sns.load_dataset("iris")
iris.head()
sns.pairplot(iris);
```

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > .15
sns.lmplot(x="total_bill", y="big_tip", data=tips, logistic=True, y_jitter=.03, ci=None)
```

### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
# 载入iris数据集
data = sns.load_dataset("iris")
data.head()
```

```python
# 萼片（sepal）和花瓣（petal）的大小关系（散点图）
data['sepal']=data.sepal_length * data.sepal_width
data['petal']=data.petal_width * data.petal_length
sns.jointplot(x='sepal',y='petal',data=data)
```

```python
# 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
df_sepal = data[['species','sepal']]
df_petal = data[['species','petal']]
# df_total = pd.concat([df_sepal,df_petal],axis=0)
df_sepal['type'] = 'sepal'
df_sepal = df_sepal.rename(columns={'sepal':'size'})
df_petal['type'] = 'petal'
df_petal = df_petal.rename(columns={'petal':'size'})
total = pd.concat([df_sepal,df_petal],axis=0)
total.sample(20)
sns.barplot(x='species',y='size',hue='type', data=total)
```

```python
# 箱式图
sns.boxplot(x='species',y='size',data=total,hue='type')
```

```python
# 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）
sns.stripplot(x='species',y='size', data=total, hue='type',jitter=True)
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
def pile_bar(data,pile_col,clsf_col):
    data1 = data.groupby([pile_col,clsf_col])['survived'].count()
    data1 = data1.unstack()
    data1 = data1/data1.sum()
    x_data = range(len(data1.columns))
    q = pd.Series([0 for i in x_data])
    bottom = q
    colors = ['#539caf', '#7663b0','darkorange']
    _,axe = plt.subplots()
    for c,p in zip(colors,data1.index):
        bottom +=q
        plt.bar(x_data, data1.loc[p],bottom=bottom,label=p,
                width=0.15*len(data1.columns),color=c)
        q = data1.loc[p].reset_index()[p]
    plt.xlabel(data1.columns.name,fontsize=20)
    plt.xticks(x_data,data1.columns, fontsize=15)
    plt.ylabel('passengers survived',fontsize=20)
    plt.legend(fontsize=15)
    axe.spines['top'].set_color(None)
    axe.spines['right'].set_color(None)
```

```python
# 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
pile_bar(data,'alive','class')
```

```python
# 不同性别的幸存比例（堆积柱状图）
pile_bar(data,'alive','sex')
```

```python
# 幸存和遇难乘客的票价分布（分类箱式图）
sns.boxplot(x='survived',y='fare',data=data)
```

```python
# 幸存和遇难乘客的年龄分布（分类箱式图）
sns.boxplot(x='survived',y='age',data=data)
```

```python
# 不同上船港口的乘客仓位等级分布（分组柱状图）
sns.barplot(x='embarked',y='survived',data=data,hue='class')
```

```python
# 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
sns.boxplot(x='alive',y='sibsp',data=data)
```

```python
# 幸存和遇难乘客父母子女的数量分布（分类箱式图）
sns.boxplot(x='alive',y='parch',data=data)
```

```python
# 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）
sns.countplot(x='alone',hue='alive',data=titanic)
```
