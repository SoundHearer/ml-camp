### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
# your code
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris) #哦，这个不区分种类
plt.show()
```

- note: 结果并不很直观 。

```python
# 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）


plt.figure(figsize=(12,8))
plt.subplot2grid((2,2),(0,0))
sns.countplot(x ="sepal_length",hue="species",data=iris)
plt.subplot2grid((2,2),(0,1))
sns.countplot(x ="petal_length",hue="species",data=iris)
plt.subplot2grid((2,2),(1,0))
sns.countplot(x ="petal_width",hue="species",data=iris)
plt.subplot2grid((2,2),(1,1))
sns.countplot(x ="sepal_width",hue="species",data=iris)
plt.show()
```

```python
# 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）
sns.pairplot(data=iris,hue="species")
plt.show()
```

### 3.餐厅小费情况分析
* 载入tips数据集
* 小费和总消费之间的关系（散点图）
* 男性顾客和女性顾客，谁更慷慨（分类箱式图）
* 抽烟与否是否会对小费金额产生影响（分类箱式图）
* 工作日和周末，什么时候顾客给的小费更慷慨（分类箱式图）
* 午饭和晚饭，哪一顿顾客更愿意给小费（分类箱式图）
* 就餐人数是否会对慷慨度产生影响（分类箱式图）
* 性别+抽烟的组合因素对慷慨度的影响（分组柱状图）

```python
# your code
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
tips.head()
```

```python
# 小费和总消费之间的关系（散点图）
sns.lmplot(x = "total_bill", y ="tip",data=tips)
plt.show()
```

```python
# 男性顾客和女性顾客，谁更慷慨（分类箱式图）
sns.boxplot(x="sex",y="tip",hue="sex",data=tips)
plt.show()
```

```python
# 抽烟与否是否会对小费金额产生影响（分类箱式图）
sns.boxplot(x="smoker",y = "tip",hue="smoker",data=tips)
plt.show()
```

```python
# 工作日和周末，什么时候顾客给的小费更慷慨（分类箱式图）
tips["weekend"] = tips["day"].isin(["Sat","Sun"])
sns.boxplot(x = "weekend",y="tip",hue="weekend",data=tips)
```

```python
# 午饭和晚饭，哪一顿顾客更愿意给小费（分类箱式图）
sns.boxplot(x = "time",y = "tip",hue = "time",data = tips)
plt.show()
```

```python
# 就餐人数是否会对慷慨度产生影响（分类箱式图）
sns.boxplot(x="size",y="tip",hue="size",data=tips)
plt.show()
```

```python
# 性别+抽烟的组合因素对慷慨度的影响（分组柱状图）
sns.barplot(x="sex", y="tip", hue="smoker",data=tips);
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
# your code
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
titanic.head()
```

- note:堆积柱状图呢?

```python
# 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
titanic["survived_pc1"] = titanic["survived"]*(titanic["pclass"] == 1)
titanic["survived_pc2"] = titanic["survived"]*((titanic["pclass"] == 2)|(titanic["pclass"] == 1))
titanic["survived_pc3"] = titanic["survived"]*((titanic["pclass"] == 3)|(titanic["pclass"] == 2)|(titanic["pclass"] == 1))
sns.countplot(x="survived_pc3",data=titanic)
sns.countplot(x="survived_pc2",data=titanic)
sns.countplot(x="survived_pc1",data=titanic)
plt.show()
# 其实我想的是，叠加三张图
```

```python
# 不同性别的幸存比例（堆积柱状图）
sns.countplot(x = "survived",hue="sex",data = titanic)
plt.show()
```

```python
# 幸存和遇难乘客的票价分布（分类箱式图）
sns.boxplot(x="survived",y="fare",data=titanic)
plt.show()
```

```python
# 幸存和遇难乘客的年龄分布（分类箱式图）
sns.boxplot(x="survived",y="age",data=titanic)
plt.show()
```

```python
# 不同上船港口的乘客仓位等级分布（分组柱状图）
sns.countplot(x="pclass",hue="embarked",data=titanic)
plt.show()
```

```python
# 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
sns.boxplot(x="survived",y="sibsp",data=titanic)
plt.show()
```

```python
# 幸存和遇难乘客父母子女的数量分布（分类箱式图）
sns.boxplot(x="survived",y="parch",data=titanic)
plt.show()
```

```python
# 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）
sns.barplot(x="survived",y="alone",data=titanic)
plt.show()
```
