***您的答案

### 4.（接上题）如何从可视化的角度解读一组数据的分布特征？(代码演示有额外加分)

***您的答案
- 散点图用来表征两个数值型变量间的关系
- 折线图用来表征两个连续数值型变量之间的关系
- 柱状图用来表征分类型变量和数值型变量的关系
- 直方图用来表征一个数值型变量的分布，横轴是数值型变量本身的值，纵轴是频数。
- 概率密度曲线图本质与直方图类似，不过纵轴不是频数而是频率
- 箱线图用来表征各个数值型变量的分布状况，每一条横线代表分位数，盒内部的横线代表中位数，点代表异常值。

### 1.航班乘客变化分析

- 载入相应绘图库及航班乘客数据（flights）

```python
# your code
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# your code
flights = sns.load_dataset('flights')
flights.head()
```

- 分析年度乘客总量变化情况（折线图）

```python
# your code
flight_groupby_year = flights[['year','passengers']].groupby('year').sum()
year = flight_groupby_year.index.values
passengers = flight_groupby_year.passengers.values
year_sum = pd.DataFrame({'year':year,'passengers':passengers},columns=['year','passengers'])
g = sns.FacetGrid(year_sum,size=8,aspect=1.8)
g.map(sns.pointplot,'year','passengers')
plt.show()
```

- 分析乘客在一年中各月份的分布（柱状图）

```python
# your code
fig,ax = plt.subplots(figsize=(14,6))
sns.barplot('month','passengers',data=flights,ax=ax)
```

### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
# your code
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['sepal_area']=iris['sepal_length']*iris['sepal_width']
iris['petal_area']=iris['petal_length']*iris['petal_width']
iris.head()
```

```python
sns.jointplot('sepal_area','petal_area',data=iris)
```

```python
g = sns.PairGrid(iris,
                 x_vars=["species"],
                 y_vars=["sepal_area", "petal_area"],
                 aspect=2, size=4)
g.map(sns.barplot, palette="Set3")
```

```python
sns.pairplot(iris[['sepal_area','petal_area','species']],hue='species')
```

### 3.餐厅小费情况分析
* 载入tips数据集
* 小费和总消费之间的关系（散点图）
* 男性顾客和女性顾客，谁更慷慨（分类箱式图）
* 抽烟与否是否会对小费金额产生影响（分类箱式图）

* 性别+抽烟的组合因素对慷慨度的影响（分组柱状图）

```python
# your code
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.boxplot(x='sex',y='tip',data=tips)
```

```python
sns.boxplot(x='smoker',y='tip',data=tips)
```

```python
def judge_weekday(x):
    if x in ('Sat','Sun'):
        return 'weekend'
    else:
        return 'weekday'
tips['weekday'] = tips.day.apply(judge_weekday)
sns.boxplot(x='weekday',y='tip',data=tips)
```

```python
sns.boxplot(x='time',y='tip',data=tips)
```

```python
sns.boxplot(x='size',y='tip',data=tips)
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
# your code
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
def died_flag(x):
    if x == 0:
        return 1
    else:
        return 0
titanic['died'] = titanic['survived'].apply(died_flag)
titanic.head()
```

```python
data_survived_pclass = titanic[['pclass','survived','died']].groupby('pclass').sum()
data_survived_pclass['sum'] = data_survived_pclass['survived'] + data_survived_pclass['died']
data_survived_pclass['survived'] = data_survived_pclass['survived']/data_survived_pclass['sum']
data_survived_pclass['died'] = data_survived_pclass['died']/data_survived_pclass['sum']
data_survived_pclass[['survived','died']].plot(kind='bar',stacked=True)
```

```python
data_survived_pclass = titanic[['sex','survived','died']].groupby('sex').sum()
data_survived_pclass['sum'] = data_survived_pclass['survived'] + data_survived_pclass['died']
data_survived_pclass['survived'] = data_survived_pclass['survived']/data_survived_pclass['sum']
data_survived_pclass['died'] = data_survived_pclass['died']/data_survived_pclass['sum']
data_survived_pclass[['survived','died']].plot(kind='bar',stacked=True)
```
