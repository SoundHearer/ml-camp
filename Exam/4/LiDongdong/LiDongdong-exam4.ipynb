### 1.谈谈您对数据可视，以及其对数据分析，机器学习的意义的理解

***通俗的可以认为，数据分析的过程应该是先进行数据的分析和理解，在对数据分析和理解的过程中需要用到可视化来辅助分析、直观的认识数据，然后在此基础上把数据的特征或者有用的特性提取出来然后用这些有用的数据进行机器学习建模，来形成我们需要的模型

### 2.某评测网站对市场上某品牌充电宝随机抽取30款测得其使用寿命数据，请问哪些可视化方案适宜呈现该结果？这些方案透露出了哪些含义

***在对该数据进行可视化分析的时候，大致需要以下几个步骤
（1）使用分类柱状图来查看不同款式的充电宝在使用寿命上大致是如何的，该方案主要是考察充电宝品牌的总体寿命情况
（2）使用分类散点图具体查看这30款充电宝其他信息对寿命的影响程度，比如电量大小、重量、电池品牌等等，该方案主要是看不同因素对整体寿命产生的了什么效果

### 3.	一组数据的分布特征可以从哪几个方面进行描述？

***分布特征可以考虑以下几个方面：
（1）数据的中位数、四分位数等，用于描述数据的整体分布和具体数据与总体数据的关系，可以识别一些离群值点
（2）数据的平均数，即数据分布中整体分布的平均状态，该值受离群值影响较大，可以反映多数情况的状态
（3）数据的众数，该值反应在该数据中绝大多数情况如何
（4）数据的标准差，该值反应数据分布中某组值的稳定程度

### 1.航班乘客变化分析

- 载入相应绘图库及航班乘客数据（flights）

```python
# your code
%matplotlib inline
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
data = sns.load_dataset("flights")
data.head()
```

- 分析年度乘客总量变化情况（折线图）

```python
# your code
#载入matplotlib资源配置参数
mpl.rc('font',size=16)
mpl.rc('axes.spines',top=False, right=False)
#按年进行分组，并计算每年乘客总量数
flight_data = data.groupby(['year'])['passengers'].sum().reset_index()
#画出乘客总量变化情况的折线图
sns.factorplot(x='year',y='passengers',data=flight_data,size=7)
plt.xticks(rotation=30)
plt.title(u"乘客总量变化情况")
plt.grid(True)
```

- 分析乘客在一年中各月份的分布（柱状图）

```python
# your code
#1、不分具体年份，汇总后的柱状图
sns.barplot(x='month',y='passengers',data=data)
```

```python
#2、将每年的具体情况展现汇总的柱状图
sns.set(color_codes=True)
sns.barplot(x='month',y='passengers',hue='year',data=data)
plt.legend(loc='upper left')
plt.ylabel('year/passengers',size=20,labelpad=10)
plt.xticks(rotation=35)
plt.xlabel('Month of Year ',size=20,labelpad=10)
```

### 2.鸢尾花花型尺寸分析

* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
# 载入iris数据集合
iris_data =  sns.load_dataset('iris')
iris_data.head()
```

```python
# 萼片（sepal）和花瓣（petal）的大小关系（散点图）
#计算sepal和petal的大小（简单认为其大小近似矩形）
iris_data.loc[:,'sepal_area'] = iris_data['sepal_length']*iris_data['sepal_width']
iris_data.loc[:,'petal_area'] = iris_data['petal_length']*iris_data['petal_width']
#画图
sns.jointplot(x='sepal_area',y='petal_area',data=iris_data)
```

```python
# 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
#柱状图
iris_data.head()
_,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,5))
#萼片分布情况
sns.barplot(x='species',y='sepal_area',data=iris_data,ax=axes[0])
#花瓣分布情况
sns.barplot(x='species',y='petal_area',data=iris_data,ax=axes[1])
```

```python
#箱式图
_2,axes2=plt.subplots(nrows=1,ncols=2,figsize=(12,5))
#萼片分布情况
sns.boxplot(x='species',y='sepal_area',data=iris_data,ax=axes2[0])
#花瓣分布情况
sns.boxplot(x='species',y='petal_area',data=iris_data,ax=axes2[1])
```

* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
#载入数据
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
fig, ax = plt.subplots(figsize = (8,5))
titanic['all']=1#设置总比例分布
sns.barplot(x = "pclass", y  = "all", data = titanic, color = "r", ax = ax)
sns.barplot(x = "pclass", y  = "survived", data = titanic, color = "g", ax = ax)
```

```python
#幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
sns.boxplot(x='survived',y='sibsp',data=titanic)
```
