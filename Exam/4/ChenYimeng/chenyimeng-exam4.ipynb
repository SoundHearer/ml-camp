### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
# 载入iris数据集
iris = sns.load_dataset('iris')
iris['sepal'] = iris['sepal_length'] * iris['sepal_width']
iris['petal'] = iris['petal_length'] * iris['petal_width']
iris.head()
```

```python
# 萼片（sepal）和花瓣（petal）的大小关系（散点图）
sns.lmplot(x = 'sepal', y = 'petal', data = iris, fit_reg = False)
```

```python
# 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）

#处理数据
sepal = iris.loc[:,['species', 'sepal']]
petal = iris.loc[:,['species', 'petal']]
sepal.columns = ['species', 'size']
sepal['kind'] = 'spetal'
petal.columns = ['species', 'size']
petal['kind'] = 'petal'
p = sepal.append(petal)

#绘制箱线图
sns.boxplot(x = 'species',y = 'size',hue = 'kind',data = p)
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
#加载数据
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）

#生成数据
# p_a = titanic.loc[:, ['pclass', 'survived', 'alive']].groupby(['pclass', 'alive']).count()
p_a_total = pd.DataFrame({'total' : titanic['pclass'].value_counts()})
p_a_alive = titanic.loc[:, ['pclass', 'survived']].groupby('pclass').sum()
p_a = p_a_total.join(p_a_alive)
#存活_比例
p_a['a_r'] = p_a['survived'] / p_a['total']
p_a = p_a.sort_index()
```

```python
#绘图
sns.barplot(x = p_a.index, y = [1] * len(p_a), data = p_a, color = 'r')
bottom_plot = sns.barplot(x = p_a.index, y = 'a_r', data = p_a, color = 'b')
topbar = plt.Rectangle((0,0),1,1,fc="red", edgecolor = 'none')
bottombar = plt.Rectangle((0,0),1,1,fc='b',  edgecolor = 'none')
l = plt.legend([bottombar, topbar], ['dead', 'survived'], loc=1, ncol = 2, prop={'size':12})
l.draw_frame(False)
sns.despine(left=True)
bottom_plot.set_xlabel('pclass')
bottom_plot.set_ylabel('ratio')
bottom_plot.set_xticklabels(p_a.index, rotation=30, fontsize=12)
bottom_plot.set_ylim (0,1.3)
plt.show()
```

```python
# 不同性别的幸存比例（堆积柱状图）

pclass = pd.DataFrame({'total': titanic['sex'].value_counts()})
p_a = titanic.loc[:, ['sex', 'survived']].groupby(['sex']).sum()
p = p_a.join(pclass)
p['survived_r'] = p['survived'] / p['total']
p['dead_r'] = 1 - p['survived'] / p['total']


sns.set_style("white")
sns.set_context({"figure.figsize": (6, 5)})
#画出整个比例柱子都为1
sns.barplot(x = p.index, y = p['survived_r'] + p['dead_r'], data = p, color = 'red')
#使用死亡比例盖住下面部分
bottom_plot = sns.barplot(x = p.index, y = p['dead_r'], data = p, color = 'b')
#画标注
topbar = plt.Rectangle((0,0),1,1,fc="red", edgecolor = 'none')
bottombar = plt.Rectangle((0,0),1,1,fc='b',  edgecolor = 'none')
l = plt.legend([bottombar, topbar], ['dead', 'survived'], loc=1, ncol = 2, prop={'size':12})
l.draw_frame(False)
sns.despine(left=True)
bottom_plot.set_xlabel('sex')
bottom_plot.set_ylabel('ratio')
bottom_plot.set_xticklabels(p.index, rotation=30, fontsize=12)
bottom_plot.set_ylim (0,1.3)
plt.show()
```
