### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
titanic_data = sns.load_dataset('titanic')
titanic_data.head()
# 幸存与否，仓位等级，性别，年龄，堂兄弟姐妹数，父母子女数，票价，上船港口缩写，仓位等级，人员分类，是否成年男性，所在甲板，上船港口，是否幸存，是否单独乘船
```

- 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）

```python
class_survived_data = titanic_data.loc[:,['class','alive']].groupby(['class','alive']).size().unstack()
class_survived_data.loc[:,'alive_percent'] = class_survived_data['yes'] / (class_survived_data['no'] + class_survived_data['yes'])
class_survived_data.loc[:,'dead_percent'] = class_survived_data['no'] / (class_survived_data['no'] + class_survived_data['yes'])
class_survived_data
```

```python
# 绘制堆积柱状图
def stackedbarplot(x_data, y_data_list, y_data_names, colors, x_label, y_label, title):
    _, ax = plt.subplots()
    # 循环绘制堆积柱状图
    for i in range(0, len(y_data_list)):
        if i == 0:
            ax.bar(x_data, y_data_list[i], color = colors[i], align = 'center', label = y_data_names[i])
        else:
            # 采用堆积的方式，除了第一个分类，后面的分类都从前一个分类的柱状图接着画
            # 用归一化保证最终累积结果为1
            ax.bar(x_data, y_data_list[i], color = colors[i], bottom = y_data_list[i - 1], align = 'center', label = y_data_names[i])
    
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.legend(loc = 'upper right') # 设定图例位置
```

```python
mpl.rc('figure', figsize = (4, 6))

# 调用绘图函数

stackedbarplot(x_data = class_survived_data.index
               , y_data_list = [class_survived_data['alive_percent'], class_survived_data['dead_percent']]
               , y_data_names = ['alive_percent', 'dead_percent']
               , colors = ['#539caf', '#7663b0']
               , x_label = 'class'
               , y_label = 'percent'
               , title = 'Survived and Dead Passangers Percent In Different Class')
```

- 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
alone_alive_data = titanic_data.loc[:,['alone','alive']].groupby(['alone','alive']).size().unstack()
alone_alive_data.loc[:,'alive_percent'] = alone_alive_data['yes'] / (alone_alive_data['no'] + alone_alive_data['yes'])
alone_alive_data.loc[:,'dead_percent'] = alone_alive_data['no'] / (alone_alive_data['no'] + alone_alive_data['yes'])
alone_alive_data
```

```python
# 绘制堆积柱状图
def stackedBool_barplot(x_data, y_data_list, y_data_names, colors, x_label, y_label, title):
    _, ax = plt.subplots()
    # 循环绘制堆积柱状图
    for i in range(0, len(y_data_list)):
        if i == 0:
            ax.bar(x_data, y_data_list[i], color = colors[i], align = 'center', label = y_data_names[i])
        else:
            # 采用堆积的方式，除了第一个分类，后面的分类都从前一个分类的柱状图接着画
            # 用归一化保证最终累积结果为1
            ax.bar(x_data, y_data_list[i], color = colors[i], bottom = y_data_list[i - 1], align = 'center', label = y_data_names[i])
    plt.xticks([False,True], [r'False', r'True'])
    ax.set_ylabel(y_label)
    ax.set_xlabel(x_label)
    ax.set_title(title)
    ax.legend(loc = 'upper right') # 设定图例位置# 调用绘图函数
    
stackedBool_barplot(x_data = alone_alive_data.index
               , y_data_list = [alone_alive_data['no'], alone_alive_data['yes']]
               , y_data_names = ['dead', 'alive']
               , colors = ['#539caf', '#7663b0']
               , x_label = 'alone'
               , y_label = 'alive count'
               , title = 'Survived and Dead Passangers Percent For Alone')
```
