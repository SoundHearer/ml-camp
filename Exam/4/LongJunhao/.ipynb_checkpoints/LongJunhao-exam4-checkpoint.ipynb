### 4.（接上题）如何从可视化的角度解读一组数据的分布特征？(代码演示有额外加分)

***您的答案 以泰坦尼克号样本举例，存活下来的女性远超男性

```python
# your code
%matplotlib inline 
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset("titanic")
titanic.head()
def pile_bar(data,pile_col,clsf_col):
    data1 = data.groupby([pile_col,clsf_col])['survived'].count()
    data1 = data1.unstack()
    data1 = data1/data1.sum()
    x_data = range(len(data1.columns))
    q = pd.Series([0 for i in x_data])
    bottom = q
    colors = ['#539caf', '#7663b0','darkorange']
    _,axe = plt.subplots()
    for c,p in zip(colors,data1.index):
        bottom +=q
        plt.bar(x_data, data1.loc[p],bottom=bottom,label=p,
                width=0.15*len(data1.columns),color=c)
        q = data1.loc[p].reset_index()[p]
    plt.xlabel(data1.columns.name,fontsize=20)
    plt.xticks(x_data,data1.columns, fontsize=15)
    plt.ylabel('passengers survived',fontsize=20)
    plt.legend(fontsize=15)
    axe.spines['top'].set_color(None)
    axe.spines['right'].set_color(None)
pile_bar(titanic,'alive','sex')
plt.title(u'不同性别的幸存比例',fontsize=20)
```

### 1.航班乘客变化分析

- 载入相应绘图库及航班乘客数据（flights）

```python
%matplotlib inline 
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
data = sns.load_dataset("flights")
data.tail()
```

- 分析年度乘客总量变化情况（折线图）

```python
allYear = data[['year','passengers']].groupby('year').sum()
allYear.reset_index(inplace=True)
sns.set(style = "ticks")
sns.factorplot(x = "year",y = "passengers",data = allYear,aspect=2)
```

- 分析乘客在一年中各月份的分布（柱状图）

```python
# your code
month_year_distribute = data.groupby('month').agg([np.mean,np.std])['passengers']
sns.factorplot(x='month',y='passengers',data = data,kind='bar',size =8)
plt.xticks(rotation = 60)
```

### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
data = sns.load_dataset("iris")
data.head()
```

```python
data['sepal_area'] = data['sepal_length']*data['sepal_width']
data['petal_area'] = data['petal_length']*data['petal_width']
data.head()
```

```python
sns.set(style = 'darkgrid')
sns.jointplot(x = "sepal_area",y = "petal_area",data = data,size = 7,ratio = 7)
```

```python
#不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
sepal_area_data = data[['species','sepal_area']]
sepal_area_data_tmp = sepal_area_data.copy()
sepal_area_data_tmp['area_type'] = 'sepal_area'
sepal_area_data_tmp.columns = ['species','area','area_type']
petal_area_data = data[['species','petal_area']]
petal_area_data_tmp = petal_area_data.copy()
petal_area_data_tmp['area_type'] = 'petal_area'
petal_area_data_tmp.columns = ['species','area','area_type']
tmpData = sepal_area_data_tmp.append(petal_area_data_tmp,ignore_index = True)
tmpData.head()
```

```python
#不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）
fig,ax = plt.subplots(figsize = (14,5))
sns.set(style = 'white')
sns.swarmplot(x = "sepal_area",y = "petal_area",hue = 'species',data = data, ax = ax)
```

```python
data['day'].unique()
```

```python
#工作日和周末，什么时候顾客给的小费更慷慨（分类箱式图）
weekday = data[['day','tip']]
weekdayTmp = weekday.copy()
weekdayTmp['is_weekend'] = weekdayTmp['day'].isin(['Sun','Sat']).replace({True:'Yes',False:'No'})
weekdayTmp.tail()
sns.boxplot(x = 'is_weekend',y = "tip",data = weekdayTmp)
```

```python
#午饭和晚饭，哪一顿顾客更愿意给小费（分类箱式图）
sns.boxplot(x = "time",y = "tip",data = data)
```

```python
#就餐人数是否会对慷慨度产生影响（分类箱式图）
sns.boxplot(x = "size",y = "tip",data = data)
```

```python
#性别+抽烟的组合因素对慷慨度的影响（分组柱状图）
sns.barplot(x = "smoker",y = "tip",hue = "sex",data = data,ci = None)
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
# your code
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
def pile_bar(data,pile_col,clsf_col):
    data1 = data.groupby([pile_col,clsf_col])['survived'].count()
    data1 = data1.unstack()
    data1 = data1/data1.sum()
    x_data = range(len(data1.columns))
    q = pd.Series([0 for i in x_data])
    bottom = q
    colors = ['#539caf', '#7663b0','darkorange']
    _,axe = plt.subplots()
    for c,p in zip(colors,data1.index):
        bottom +=q
        plt.bar(x_data, data1.loc[p],bottom=bottom,label=p,
                width=0.15*len(data1.columns),color=c)
        q = data1.loc[p].reset_index()[p]
    plt.xlabel(data1.columns.name,fontsize=20)
    plt.xticks(x_data,data1.columns, fontsize=15)
    plt.ylabel('passengers survived',fontsize=20)
    plt.legend(fontsize=15)
    axe.spines['top'].set_color(None)
    axe.spines['right'].set_color(None)
```

```python
plt.rcParams['font.sans-serif']=['SimHei']
pile_bar(titanic,'alive','class')
plt.rcParams["figure.figsize"] = (12,6)
plt.title(u'不同仓位等级中幸存和遇难的乘客比例',fontsize=20)
```

```python
pile_bar(titanic,'alive','sex')
plt.title(u'不同性别的幸存比例',fontsize=20)
```
