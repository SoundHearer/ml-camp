### 1.航班乘客变化分析

- 载入相应绘图库及航班乘客数据（flights）

```python
# your code
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="ticks", color_codes=True)
fly = sns.load_dataset("flights")
fly.head()
```

- 分析年度乘客总量变化情况（折线图）

```python
# your code
tmp = fly.groupby(['year'])['passengers'].sum()#这样分类是一个series
tmp2=pd.DataFrame(tmp)#转化成dataframe
tmp2.reset_index(inplace=True)#重新定义index，去掉year作为index，也就是year是数据，不是index
tmp2.head()#年度乘客总量
```

```python
f_p=sns.pointplot(x='year',y='passengers',data=tmp2)#
f_p.set_ylabel('Pasengers')
```

- 分析乘客在一年中各月份的分布（柱状图）

```python
# your code
month_data=pd.DataFrame(fly.groupby('month')['passengers'].agg(sum))#按照月份分类
month_data.reset_index(inplace=True)
fig,ax=plt.subplots(figsize = (12,6))#设置图片长宽
sns.barplot(x='month',y='passengers',data=month_data,ax=ax)
```

### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
# your code
#载入iris数据集
iris = sns.load_dataset("iris")
iris.head()
```

```python
# 萼片（sepal）和花瓣（petal）的大小关系（散点图）
'''先计算出萼片面积和花瓣面积'''
iris['sepal_size']=iris['sepal_length']*iris['sepal_width']
iris['petal_size']=iris['petal_length']*iris['petal_width']
iris.head()
```

```python
# 萼片（sepal）和花瓣（petal）的大小关系（散点图）
sns.jointplot(x="sepal_size",y="petal_size",data=iris)
```

```python
# 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）
sns.pairplot(iris,hue='species')#hue参数可以实现分类
```

```python
# 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
'''首先取出不同种类鸢尾花萼片和花瓣大小的数据'''
t1=iris[['species','sepal_size']]#种类和萼片大小的数据
t1['size_type']='sepal_size'
t1.rename(columns={'sepal_size':'size', }, inplace = True)
# t1.head()
t2=iris[['species','petal_size']]#种类和花瓣大小的数据
t2['size_type']='petal_size'
t2.rename(columns={'petal_size':'size'}, inplace = True)
# t2.head()
tmp=pd.concat([t1,t2],axis=0)
tmp.head()#所有画图是根据列名中的数据去分类画图的，因此，需要把花萼和花瓣大小作为一列数据，另外再设置一列大小类型的说明来区分花萼和花瓣大小
```

```python
# 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
i_p=sns.barplot(x='species',y='size',hue='size_type',data=tmp)
i_p.set_ylabel('size')
```

```python
#  不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
sns.boxplot(x='species',y='size',hue='size_type',data=tmp)
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
# your code
tic = sns.load_dataset("titanic")
tic.head()
```

```python
# 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
'''首先对class，alive数据进行处理'''
tic1=tic[['class','alive']].groupby(['class','alive']).size().unstack()#分类、计数、分解
tic1.reset_index(inplace=True)
tic1['total']=tic1['no']+tic1['yes']#增加一列方便画图
tic1#不同仓位等级中幸存和遇难的乘客数据表格
```

```python
total_plot=sns.barplot(x='class',y='total',data=tic1,color='r')#先画总人数的柱状图，因为总数柱子长(这是一个ax中的底层)
sns.barplot(x='class',y='yes',data=tic1,color='b')#再画生还人数的柱状图(这是一个ax中的第二层)，下面的图其实就是两层按照X,Y轴重叠起来的，红色部分减去被蓝色部分覆盖的就是没有生还的人数
alive_Y=plt.Rectangle((0,0),1,1,fc='b', edgecolor = 'none')#图例的表示图案
alive_N=plt.Rectangle((0,0),1,1,fc='r', edgecolor = 'none')
plt.legend([alive_Y, alive_N], ['alive_Y', 'alive_N'], loc='best', ncol = 2, prop={'size':16})#图例
total_plot.set_ylabel('passengers total')#给Y轴加上标签说明
```

```python
# 不同性别的幸存比例（堆积柱状图）
tic2=tic[['sex','alive']].groupby(['sex','alive']).size().unstack()
tic2.reset_index(inplace=True)
tic2['total']=tic2['no']+tic2['yes']
tic2#不同性别的幸存和遇难人数表格
```

```python
total_plot=sns.barplot(x='sex',y='total',data=tic2,color='r')
sns.barplot(x='sex',y='yes',data=tic2,color='b')
alive_Y=plt.Rectangle((0,0),1,1,fc='b', edgecolor = 'none')
alive_N=plt.Rectangle((0,0),1,1,fc='r', edgecolor = 'none')
plt.legend([alive_Y, alive_N], ['alive_Y', 'alive_N'], loc='upper left', ncol = 2, prop={'size':16})
total_plot.set_ylabel('passengers total')
```

```python
# 幸存和遇难乘客的票价分布（分类箱式图）
sns.set_context({"figure.figsize": (14, 12)})#设置图片大点
sns.boxplot(x='alive',y='fare',data=tic)
```

```python
# 幸存和遇难乘客的年龄分布（分类箱式图）
sns.set_context({"figure.figsize": (10, 8)})#设置小点
sns.boxplot(x='alive',y='age',data=tic)
```

```python
# 不同上船港口的乘客仓位等级分布（分组柱状图）
'''对一，二 三 等仓的港口数据分别处理，然后汇总成一张dataframe'''
E1=pd.DataFrame(tic[tic.pclass==1]['embarked'])#一等舱港口的乘客数据
e1=pd.DataFrame(E1.groupby('embarked').size())
e1.reset_index(inplace=True)
e1.columns=['embarked','numbers']
e1['em_class']='First'
# e1

E2=pd.DataFrame(tic[tic.pclass==2]['embarked'])#二等舱港口的乘客数据
e2=pd.DataFrame(E2.groupby('embarked').size())
e2.reset_index(inplace=True)
e2.columns=['embarked','numbers']
e2['em_class']='Second'
# e2
E3=pd.DataFrame(tic[tic.pclass==3]['embarked'])#三等舱港口的乘客数据
e3=pd.DataFrame(E3.groupby('embarked').size())
e3.reset_index(inplace=True)
e3.columns=['embarked','numbers']
e3['em_class']='Third'
# e3

tic5=pd.concat([e1,e2,e3],axis=0)
tic5#各个仓位和各个港口的乘客数量表格
```

```python
em_plot=sns.barplot(x='embarked',y='numbers',hue='em_class',data=tic5)
em_plot.set_ylabel('Passengers')
```

```python
# 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
sns.boxplot(x='survived',y='sibsp',data=tic)
```

```python
# 幸存和遇难乘客父母子女的数量分布（分类箱式图）
sns.boxplot(x='survived',y='parch',data=tic)
```

```python

# 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）
tic9=tic[['alone','survived']].groupby(['alone','survived']).size().unstack()
tic9.reset_index(inplace=True)
tic9.columns=['alone','alive_N','alive_Y']
tic9['total']=tic9['alive_N']+tic9['alive_Y']
tic9#单独乘船与否和幸存与否的数据表格
```

```python
# 单独乘船与否和幸存之间有没有联系（堆积柱状图）
sns.barplot(x='alone',y='total',data=tic9,color='r',ci=None)
alone_plot=sns.barplot(x='alone',y='alive_Y',data=tic9,color='b',ci=None)
alive_Y=plt.Rectangle((0,0),1,1,fc='b', edgecolor = 'none')#图例的表示图案
alive_N=plt.Rectangle((0,0),1,1,fc='r', edgecolor = 'none')
plt.legend([alive_Y, alive_N], ['alive_Y', 'alive_N'], loc='upper left', ncol = 2, prop={'size':16})#图例
alone_plot.set_ylabel('Passengers')
```

```python
tic91=tic9[['alone','alive_N']]
tic91['alive']='no'
tic91.rename(columns={'alive_N':'numbers'},inplace=True)
# tic91
tic92=tic9[['alone','alive_Y']]
tic92['alive']='yes'
tic92.rename(columns={'alive_Y':'numbers'},inplace=True)
# tic92
tic10=pd.concat([tic91,tic92],axis=0)
tic10##单独乘船与否和幸存与否的数据表格
```

```python
# 单独乘船与否和幸存之间有没有联系（分组柱状图）
sns.barplot(x='alone',y='numbers',hue='alive',data=tic10)
```
