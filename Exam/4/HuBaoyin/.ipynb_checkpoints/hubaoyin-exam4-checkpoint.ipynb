### 1.谈谈您对数据可视，以及其对数据分析，机器学习的意义的理解

答：数据可视化是指使用图形化的手段展示数据，从数据提取关键特征和模式，数据可视化能够让数据分析更加直观、简便，提高数据分析的效率。要进行机器学习建模数据分析和特征的抽取工作必不可少，数据和特征决定了模型发挥作用的上限，数据可视化能够让建模人员快速抓住数据的有效特征和变化趋势，从而提高机器学习建模的效率，提高模型的效果。

### 2.某评测网络对市场上某品牌充电宝随机抽取30款测得其使用寿命数据，请问哪些可视化方案适宜呈现该结果？这些方案透露出了哪些含义

答：
1.方案： 柱状图   含义：不同品牌之间使用寿命对比，从柱形的高低可以直接看出不同充电宝的使用寿命对比情况；
2.方案： 箱线图   含义：30个品牌充电宝使用寿命的中位数、%25分位线、%75分位线、离群值或者异常值

### 3.	一组数据的分布特征可以从哪几个方面进行描述？

答：走势、最大值、最小值、方差、平均值（期望）、标准差、中位数、占比、总数、空值个数、非空值个数

### 4.（接上题）如何从可视化的角度解读一组数据的分布特征？(代码演示有额外加分)

1.解读数据分布

答：单个变量分布情况可以使用灰度图distplot；两个变量可以使用散点图jointplot；均值可以使用柱状图barplot；
中位数、离群值可以使用箱线图boxplot；数据点直接展示可以使用蜂群图swarmplot；数据的变化趋势可以使用点图pointplot

2.代码演示

2.1灰度图distplot

```python
import numpy as np
sns.distplot(np.random.normal(size=200), kde=True, bins=20)
```

2.2散点图jointplot

```python
df = pd.DataFrame(np.random.multivariate_normal([0, 1], [(1, .5), (.5, 1)], 200), columns=["x", "y"])
sns.jointplot(x="x", y="y", data=df)
```

2.3柱状图barplot

```python
titanic = sns.load_dataset("titanic")
sns.barplot(x="survived", y="age", hue="class", data=titanic)
```

2.4箱线图boxplot

```python
sns.boxplot(x="survived", y="age", hue="pclass", data=titanic)
```

2.5蜂群图swarmplot

```python
sns.swarmplot(x="survived", y="age",hue='sex', data=titanic)
```

2.6点图pointplot

```python
sns.pointplot(x="pclass", y="age", data=titanic)
```

### 1.航班乘客变化分析

- 载入相应绘图库及航班乘客数据（flights）

```python
flights_data = sns.load_dataset("flights")
# 年份，月份，乘客数
year_change = flights_data.groupby(['year'])['passengers'].sum().reset_index()
```

- 分析年度乘客总量变化情况（折线图）

```python
sns.pointplot(year_change['year'], year_change['passengers'] ,\
              color='green', linestyle='-', marker='*')
```

- 分析乘客在一年中各月份的分布（柱状图）

```python
month_change = flights_data.groupby(['month'])['passengers'].sum().reset_index()
sns.barplot(x="month", y="passengers", data=month_change)
```

### 2.鸢尾花花型尺寸分析

* 载入iris数据集
* 萼片（sepal）和花瓣（petal）的大小关系（散点图）
- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）
* 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

- 载入iris数据集

```python
iris_data = sns.load_dataset("iris")
```

- 萼片（sepal）和花瓣（petal）的大小关系（散点图）

```python
iris_data.loc[:,'speal_area'] = iris_data['sepal_length']*iris_data['sepal_width']
iris_data.loc[:,'petal_area'] = iris_data['petal_length']*iris_data['petal_width']
sns.jointplot(x='speal_area', y='petal_area', data=iris_data)
```

- 不同种类鸢尾花萼片和花瓣大小的分布情况（柱状图或者箱式图）

```python
petal_tmp = iris_data[['petal_area','species']].rename(columns={'petal_area': 'area'})
petal_tmp.loc[:,'type'] = 'petal'
speal_tmp = iris_data[['speal_area','species']].rename(columns={'speal_area': 'area'})
speal_tmp.loc[:,'type'] = 'speal'
iris_box_data =pd.concat([petal_tmp,speal_tmp])
sns.boxplot(x='species', y='area', hue='type', data=iris_box_data)
```

- 不同种类（species）鸢尾花萼片和花瓣的大小关系（分类散点子图）

```python
sns.stripplot(x="speal_area", y="petal_area",hue='species', data=iris_data)
```

### 4.泰坦尼克号海难幸存状况分析
* 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）
* 不同性别的幸存比例（堆积柱状图）
* 幸存和遇难乘客的票价分布（分类箱式图）
* 幸存和遇难乘客的年龄分布（分类箱式图）
* 不同上船港口的乘客仓位等级分布（分组柱状图）
* 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）
* 幸存和遇难乘客父母子女的数量分布（分类箱式图）
* 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

- 载入tatanic数据集

```python
titanic_data = sns.load_dataset("titanic")
```

- 不同仓位等级中幸存和遇难的乘客比例（堆积柱状图）

```python
pclass_survived = titanic_data[['survived', 'pclass']].groupby(['pclass','survived']).size().to_frame().reset_index()
pclass_survived.rename(columns={0:'survived_count'},inplace=True)
pclass = titanic_data[['survived', 'pclass']].groupby(['pclass']).size().to_frame().reset_index()
pclass.rename(columns={0:'class_total_count'},inplace=True)
pclass_survived = pd.merge(pclass_survived,pclass,on='pclass',how='left')
pclass_survived.loc[:,'prob'] = pclass_survived['survived_count']/pclass_survived['class_total_count']
pclass_survived.loc[:,'total_prob'] = 1
```

```python
fig, ax = plt.subplots(figsize = (11,5)) #设置图的大小
sns.barplot(x = "pclass", y  = "total_prob", data = pclass_survived, color = "r", ax = ax)
sns.barplot(x = "pclass", y  = "prob", data = pclass_survived.loc[pclass_survived['survived']==1], color = "b", ax = ax)
survived1_bar = plt.Rectangle((0,0),1,1,fc='b',  edgecolor = 'none')
survived0_bar = plt.Rectangle((0,0),1,1,fc='r',  edgecolor = 'none')
plt.legend([survived1_bar, survived0_bar], ['survived', 'unsurvived'], ncol = 1, prop={'size':12},loc = 'upper center')
#图例不覆盖图表
plt.ylim(0, 1.3)
```

- 不同性别的幸存比例（堆积柱状图）

```python
sex_survived = titanic_data[['sex', 'survived']].groupby(['sex','survived']).size().to_frame().reset_index()
sex_survived.rename(columns={0:'survived_count'},inplace=True)
sex = titanic_data[['survived', 'sex']].groupby(['sex']).size().to_frame().reset_index()
sex.rename(columns={0:'sex_total_count'},inplace=True)
sex_survived = pd.merge(sex_survived,sex,on='sex',how='left')
sex_survived.loc[:,'prob'] = sex_survived['survived_count']/sex_survived['sex_total_count']
sex_survived.loc[:,'total_prob'] = 1

fig, ax = plt.subplots(figsize = (11,5)) #设置图的大小
sns.barplot(x = "sex", y  = "total_prob", data = sex_survived, color = "r", ax = ax)
sns.barplot(x = "sex", y  = "prob", data = sex_survived.loc[sex_survived['survived']==1], color = "b", ax = ax)
survived1_bar = plt.Rectangle((0,0),1,1,fc='b',  edgecolor = 'none')
survived0_bar = plt.Rectangle((0,0),1,1,fc='r',  edgecolor = 'none')
plt.legend([survived1_bar, survived0_bar], ['survived', 'unsurvived'], ncol = 1, prop={'size':12},loc = 'upper center')
#图例不覆盖图表
plt.ylim(0, 1.3)
```

- 幸存和遇难乘客的票价分布（分类箱式图）

```python
sns.boxplot(x='survived', y='fare', data=titanic_data)
```

- 幸存和遇难乘客的年龄分布（分类箱式图）

```python
sns.boxplot(x='survived', y='age', data=titanic_data)
```

- 不同上船港口的乘客仓位等级分布（分组柱状图）

```python
sns.countplot(x="embarked", hue="pclass", data=titanic_data)
```

- 幸存和遇难乘客堂兄弟姐妹的数量分布（分类箱式图）

```python
sns.boxplot(x='survived', y='sibsp', data=titanic_data)
```

- 幸存和遇难乘客父母子女的数量分布（分类箱式图）

```python
sns.boxplot(x='survived', y='parch', data=titanic_data)
```

- 单独乘船与否和幸存之间有没有联系（堆积柱状图或者分组柱状图）

```python
sns.countplot(x="alone", hue="survived", data=titanic_data)
```
