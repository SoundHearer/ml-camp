## 1. 项目需求
### 1.1 项目描述
当你和一个房屋买家讨论他们心中梦想的房屋时，一开始的时候他们不太会讨论地下室天花板的高度或者房屋与东西铁路的接近程度。然而这次挑战的数据集提供了比卧室数量或者白色栅栏数量还要多得多的价格影响因素。这里有79个可解释的变量用于预测Ames, Iowa的房价水平，它们几乎描述了住宅的各个方面，现在需要通过这些因素来预测最终的房价。
你可能需要：<br/>
 - 富有创造力的特征工程
 - 采用高级的回归算法，比如梯度提升，随机森林
 
### 1.2 目标
预测每间房屋的售价,测试集里的每一个id都要预测出一个售价.
It is your job to predict the sales price for each house. For each Id in the test set, you must predict the value of the SalePrice variable. 
### 1.3 评判标准Metric

提交的结果会通过均方对数误差来评价, 这样贵的房屋和便宜的房屋对误差的影响就会接近.

### 1.4 提交格式
提交的文件应该具有如下格式:

Id,SalePrice <br>
1461,169000.1 <br>
1462,187724.1233 <br>
1463,175221 <br>
etc.

## 2.数据处理与特征工程

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
```

```python
train_raw = pd.read_csv("train.csv")
test_raw = pd.read_csv("test.csv")
```

```python
train_raw.head()
```

```python
train_raw.columns
```

- 这里不需要Id值

```python
train_raw = train_raw.drop(columns="Id")

test_id = test_raw["Id"]
test_raw = test_raw.drop(columns="Id")
```

- 需要检查一下数据类型, 值类型和枚举类型需要分开处理, 从原始数据来看, 数值类型只有有限的项.

```python
test_raw.get_dtype_counts()
```

```python
train_raw.get_dtype_counts()
```

可以发现其中有43项不是数值类型,考虑用One-hot编码, 为了处理方便, 把数值型全部转成float32型

```python
data_types = train_raw.dtypes
numerics_cols = data_types.loc[data_types != object].index
enum_cols = data_types.loc[data_types == object].index

numerics_cols = numerics_cols[:-1]
```

```python
for col in numerics_cols:
    train_raw[col] = train_raw[col].astype("float32")
    test_raw[col] = test_raw[col].astype("float32")
```

```python
train_raw[numerics_cols].describe()
```

- 做图看看数值型的取值分布

```python
sns.distplot(train_raw["SalePrice"])

for col in numerics_cols:
    sns.jointplot(x=col,y="SalePrice", data=train_raw)
```

- 根据图像观察, GrLiveArea 和房价相关性较高, 需要抛掉一些异常值 

```python
train_raw = train_raw.iloc[ np.logical_not(np.logical_and(train_raw["GrLivArea"] > 4000, train_raw["SalePrice"] < 300000).values)]
```

```python
y_train = train_raw["SalePrice"]
train_raw = train_raw.drop(columns="SalePrice")
```

```python
split_at = train_raw.shape[0]
all_data = pd.concat([train_raw,test_raw])
```

- 查看一下缺失值的情况, 这里只查看数值型的缺失值, 枚举型的缺失值就代表"没有"这一种类别

```python
_, ax = plt.subplots(figsize=(18,6))
sns.barplot(x=numerics_cols,y = all_data[numerics_cols].isnull().apply(np.sum), ax= ax)
plt.xticks(rotation='90');
```

```python
all_data[numerics_cols].apply(lambda x: np.sum(pd.isna(x)))
```

- 可以发现LotFrontage的缺失值太多了, 可以用邻居的值来填充.

```python
all_data["LotFrontage"] = all_data.groupby("Neighborhood")["LotFrontage"].transform(lambda x:x.fillna(x.mean()))
```

- 剩下的用0填充

```python
for col in numerics_cols:
    all_data[col] = all_data[col].fillna(0)
```

- 参考[kernel](https://www.kaggle.com/serigne/stacked-regressions-top-4-on-leaderboard) 用Box-Cox对skewness较大的变量做变换

```python
from scipy.stats import norm, skew
from scipy.special import boxcox1p
```

```python
skewness = all_data[numerics_cols].apply(skew)
skewness
```

```python
highly_skew = skewness.index[np.abs(skewness) > 1]
highly_skew
```

```python
lam = 0.15
for col in highly_skew:
    #all_data[feat] += 1
    all_data[col] = boxcox1p(all_data[col], lam)
```

- 进行One-hot编码

```python
all_data_one_hot = all_data
for c in enum_cols:
    if np.sum(all_data_one_hot[c].apply(lambda x:pd.isna(x))) == 0:
       all_data_one_hot = pd.get_dummies(all_data_one_hot, columns=[c])
    else:
        all_data_one_hot = pd.get_dummies(all_data_one_hot, columns=[c], dummy_na=True)
```

```python
all_data_one_hot.shape
```

- 看看还没有没有Nan

```python
np.sum(pd.isna(all_data_one_hot).values)
```

```python
train_data = all_data_one_hot[:split_at]
test_data = all_data_one_hot[split_at:]
```

```python
train_data.shape
```

### Check point

```python
train_data.to_csv("train_processed.csv")
test_data.to_csv("test_processed.csv")
y_train.to_csv("train_label.csv")
```

## 3.训练模型

这是一个回归问题, 可考虑采用xgboost来训练模型, 训练时采取对房价去对数的值作为监督信息, 这样可以压缩房价的幅度, 提高范化能力.

```python
from sklearn.model_selection import train_test_split

X = pd.read_csv("train_processed.csv")
test_data = pd.read_csv("test_processed.csv")

y = np.log1p(pd.read_csv("train_label.csv", header=None)[1])
```

```python
#划分训练集, 测试集
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state=20180409 )
```

```python
# 自定义的均方对数误差
def rmse(predict,target):
    return np.sqrt(np.mean((predict - target)**2))

#生成提交文件
def write_result(filename,res):
    result = pd.DataFrame()
    result["Id"] = test_id
    result["SalePrice"] = res
    result.to_csv(filename,index=False)
  
#交叉验证
from sklearn.model_selection import KFold
def cv_score(n,model,X,y):
    score = 0
    kf = KFold(n_splits=n)
    for train_idx, test_idx in kf.split(X):
        model.fit(X.values[train_idx],y.values[train_idx])
        score += rmse(model.predict(X.values[test_idx]),y.values[test_idx])

    print(score/n)
    return score/n
```

```python
import xgboost as xgb
```

Xboost的参数选择参考了Kernel: https://www.kaggle.com/serigne/stacked-regressions-top-4-on-leaderboard

在他的基础上进行了微调, 使得验证集上误差进一步下降.

```python
trees = xgb.XGBRegressor(max_depth = 3, n_estimators = 900,subsample=0.4,
                         learning_rate = 0.02, njobs = 8, min_child_weight = 4.0,
                         booster="gbtree",reg_lambda=0.8,reg_alpha=0.5)
```

```python
trees.fit(X_train, y_train)
print("train_score",rmse(trees.predict(X_train),y_train))
print("test_score",rmse(trees.predict(X_test),y_test))
```

- 参数确定下来后用全部训练集进行训练

```python
trees.fit(X, y)
cv_score(6,trees,X,y)

res = trees.predict(test_data.values)
```

```python
write_result("result1.csv",np.expm1(res))
```

- 提交后得到测试集均方对数误差 0.12466

- 可以发现, One-hot编码后数据维度有300维, 而训练数据只有1400多条, 训练集数据较小, 这使得从训练数据本身就无法准确估计整体分布. 提升树一类的模型拟合能力较强, 这使得在小数据集上过拟合会比较严重, 于是在训练过程中采用早停法, 当验证集上误差不再下降时停止增加集成规模. 通过对测试集的预测结果做图发现, 训练集上的房价分布和测试集上的房价分布不同, 这表明在测试集上的误差会高于验证集.

- 尝试了神经网络，数据集太小了，过拟合很严重, 输出也很不稳定.

- 参考其他kernel之后发现可以用Box-Cox对skewness较大的特征进行变换, 于是在之前的数据处理过程中加入skewness的计算以及Box-Cox变换.

- 提交后分数为0.12204, 比之前的0.12466有所提升.

由于L1正则化能过滤稀疏数据中的噪声, 于是尝试用带L1正则化的Lasso模型来拟合.

```python
from sklearn.linear_model import Lasso
```

```python
#参数搜索
score = 20;
for a in np.linspace(0.0001,0.0007,20):

    lasso = Lasso(alpha=a, max_iter=2500)
    lasso.fit(X,y)
    #res3 = knn.predict(test_data)
    score_new = cv_score(8,lasso,X,y)
    if score>score_new:
        score = score_new
        p = a
```

```python
print(p)
lasso = Lasso(alpha=p, max_iter=2500)
lasso.fit(X,y)
```

```python
res2 = lasso.predict(test_data)
write_result("result2.csv",np.expm1(res2))
```

## 4.总结

整个比赛做下来, 发现模型效果好不好, 特征工程很重要, 做训练之前往往要花很多时间来处理特征, 特征好不好决定了范化能力的上界. 模型选择也很重要, 深度神经网络拟合能力太强, 在这种小数据集上过拟合很严重, 输出波动很大, 本次数据稀疏性很强, 反而适合一些带L1正则化的简单模型. 再一个调参过程很麻烦, 有些模型参数非常多, 就很难调. 而且模型融合往往有提升效果, 使用较为广泛, 但是值得指出一点, 线性模型和线性模型叠加基本没用.
