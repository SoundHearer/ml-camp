```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
data_train = pd.read_csv('data/train.csv')
data_train.head()
```

### 1.label分布情况

```python
from scipy import stats
from scipy.stats import norm
sns.distplot(data_train.SalePrice,fit=norm)
#基本符合正态分布
```

```python
res = stats.probplot(data_train['SalePrice'],plot=plt)
```

```python
#取log看看
data_train['SalePrice'] = np.log(data_train['SalePrice'])
sns.distplot(data_train.SalePrice,fit=norm)
fig = plt.figure()
res = stats.probplot(data_train['SalePrice'],plot=plt)
```

### 2.label与numercical variables 关系

```python
var = 'GrLivArea'
data = pd.concat([data_train['SalePrice'],data_train[var]],axis=1)
data.plot.scatter(x=var,y='SalePrice',ylim=(0,800000))
#呈现线性分布
```

```python
#删除可能噪音
data_train.sort_values(by = 'GrLivArea',ascending=False)[:2]
```

```python
data_train = data_train.drop(data_train[data_train['Id']==1299].index,axis=0)
data_train = data_train.drop(data_train[data_train['Id']==524].index,axis=0)
```

```python
#取log看看
data_train['GrLivArea'] = np.log(data_train['GrLivArea'])
sns.distplot(data_train.SalePrice,fit=norm)
fig = plt.figure()
res = stats.probplot(data_train['GrLivArea'],plot=plt)
```

```python
var = 'TotalBsmtSF'
data = pd.concat([data_train['SalePrice'],data_train[var]],axis=1)
data.plot.scatter(x=var,y='SalePrice',ylim=(0,800000))
#呈现线性分布,do nothing
```

```python
#0值不能取log，新创建1列，if area>0,gets 1;if area=0,gets 0
data_train['HasBsmt'] = 0
data_train.loc[data_train['TotalBsmtSF']>0,'HasBsmt'] = 1
```

```python
data_train.loc[data_train['HasBsmt']==1,'TotalBsmtSF'] = np.log(data_train['TotalBsmtSF'])
```

```python
sns.distplot(data_train[data_train['TotalBsmtSF']>0]['TotalBsmtSF'],fit=norm)
fig = plt.figure()
res = stats.probplot(data_train[data_train['TotalBsmtSF']>0]['TotalBsmtSF'],plot=plt)
```

### 2.label与categorical feature 关系

```python
var = 'OverallQual'
data = pd.concat([data_train['SalePrice'],data_train[var]],axis=1)
_,ax = plt.subplots(figsize=(8,6))
fig = sns.boxplot(x=var,y='SalePrice',data=data)
fig.axis(ymin=0,ymax=800000)
#线性相关，且越大方差越大
```

```python
var = 'YearBuilt'
data = pd.concat([data_train['SalePrice'],data_train[var]],axis=1)
_,ax = plt.subplots(figsize=(20,12))
fig = sns.boxplot(x=var,y='SalePrice',data=data)
fig.axis(ymin=0,ymax=800000)
plt.xticks(rotation=90)
#总体来看，最近几年相对贵于old old年
```

```python
corrmat = data_train.corr()
f,ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat,vmax=.8,square=True)
#分析各特征的相关性(只能数值型！)
```

```python
import numpy as np
#分析前10个数值特征与SalePrice相关性
k = 10
cols = corrmat.nlargest(k,'SalePrice')['SalePrice'].index
cm = np.corrcoef(data_train[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm,cbar=True,annot=True,fmt='.2f',yticklabels=cols.values,xticklabels=cols.values)
plt.show()
```

```python
cols = corrmat.nlargest(k,'SalePrice')['SalePrice'].index
cols
```

```python
#取其中6个
cols=['SalePrice','OverallQual','GrLivArea','GarageCars','TotalBsmtSF','FullBath','YearBuilt']
sns.pairplot(data_train[cols],size=2.5)
plt.show()
```

## missing data

```python
total = data_train.isnull().sum().sort_values(ascending=False)
percent = (data_train.isnull().sum()/data_train.isnull().count()).sort_values(ascending=False)
miss_data = pd.concat([total,percent],axis=1,keys=['Total','Percent'])
miss_data.head(20)
```

```python
#删除缺失值大于1的列，删除缺失值=1相应行
data_train = data_train.drop((miss_data[miss_data['Total']>1].index),axis=1)
data_train = data_train.drop(data_train.loc[data_train['Electrical'].isnull()].index,axis=0)
data_train.isnull().sum().max()
```

```python
# dummy variables
data_train = pd.get_dummies(data_train)
```
