### 6.总结
1. 不足之处：
    - 处理缺失值数据时，简单地把缺失值>1的特征全部删除，可能会失去一些重要特征。
    - 在对数据进行可视化时，应该对类别型特征进行可视化，以便挖掘出更多更有意义的特征。
    - 在特征工程、特征选择环节多做些工作，比如尝试一些统计特征、多项式特征、进行特征重要度排序筛选等。
    - 可以多尝试几种模型，比如SVM、DNN等。
    - 模型融合在回归问题中效果良好，应该尝试多尝试，比如抽取特征子集的xgboost模型融合、不同模型融合等。
    - 在模型训练完成后，没有充分考虑过拟合现象直接去预测测试集，应该先考察模型是否过拟合；如果过拟合了添加正则化项是个很好的选择。
2. 教训：
    - 特征处理时一定要保证训练集和测试集特征的一致性，保证训练集和label的对应关系。
    - 在对类别特征进行独热向量编码时，最好拼接训练集和测试集一起进行；但如果要删除样本，只能从训练集中删除。
3. 本项目心得：
    - 模型不是重点，重要的是数据。
    - 在数据易获得的前提下，特征处理至关重要。
    - 对实际问题的理解在数据处理过程中将产生深远影响。
