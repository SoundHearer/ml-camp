### 3、日期类型

日期字段 <br>
GarageYrBlt: Year garage was built<br>
车库建成 年份  有空值<br>
YearBuilt: Original construction date<br>
建造日期<br>
YearRemodAdd: Remodel date<br>
改建日期<br>

```python
# 月销售与年销售  是日期
# YrSold
# MoSold
```

```python
# 树模型里面日期不用转为object，直接使用数字
year_process_fun = lambda x: pd.datetime.strptime(str(x), '%Y').strftime('%Y')
month_process_fun = lambda x: pd.datetime.strptime(str(x), '%m').strftime('%m')

y_m_process_fun = lambda x: pd.datetime.strptime(str(x), '%Y%m').strftime('%Y-%m')
```

```python
X['YearBuilt']  = X['YearBuilt'].astype(int).map(year_process_fun)
X['YearRemodAdd']  = X['YearRemodAdd'].astype(int).map(year_process_fun)
X['YrSold']  = X['YrSold'].astype(int).map(year_process_fun)

X['MoSold']  = (X['MoSold']).astype(int).map(month_process_fun)

# X['YMSold']  = (X['YrSold'] + X['MoSold']).map(y_m_process_fun)
```

```python
test['YearBuilt']  = test['YearBuilt'].astype(int).map(year_process_fun)
test['YearRemodAdd']  = test['YearRemodAdd'].astype(int).map(year_process_fun)
test['YrSold']  = test['YrSold'].astype(int).map(year_process_fun)

test['MoSold']  = (test['MoSold']).astype(int).map(month_process_fun)

# test['YMSold']  = (test['YrSold'] + test['MoSold']).map(y_m_process_fun)
```

### 6、预测

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
from sklearn.metrics import make_scorer, mean_squared_error
```

```python
def mean_squared_error_(ground_truth, predictions):
    return mean_squared_error(ground_truth, predictions) ** 0.5
RMSE = make_scorer(mean_squared_error_, greater_is_better=False) 
```

```python
rfr = RandomForestRegressor(n_jobs=1, random_state=0)
param_grid = {'n_estimators': [1500, 2000], 'max_features': [15,25,35], 'max_depth':[11,15,25]}
model = GridSearchCV(estimator=rfr, param_grid=param_grid, n_jobs=1, cv=10, scoring=RMSE)
model.fit(X_train, y_train)
```

```python
print('Random forecast regression...')
print('Best Params:')
print(model.best_params_)
print('Best CV Score:')
print(-model.best_score_)
score = -model.best_score_
y_pred = model.predict(test_data)
```

```python

now = datetime.datetime.now()
sub_file = 'submission/submission_'+str(score)+'_'+str(now.strftime("%Y-%m-%d-%H-%M"))+'.csv'
print ('Creating submission: ', sub_file)
pd.DataFrame({'Id': test['Id'].values, 'SalePrice': np.exp(y_pred)}).to_csv(sub_file, index=False)
```
