## 评测标准

```python
def rmse_error_(ground_truth, predictions):
    return mean_squared_error(ground_truth, predictions) ** 0.5
RMSE = make_scorer(rmse_error_, greater_is_better=False) 
```

```python
def XGB_native(train_data,test_data,features,features_non_numeric,test,y):
    depth = 3
    eta = 0.01
    ntrees = 8000
    mcw = 5
    params = {"objective": "reg:linear",
              "booster": "gbtree",
              "eta": eta,
              "max_depth": depth,
              "min_child_weight": mcw,
              "subsample": 0.5,
              "colsample_bytree": 0.5,
              "silent": 1
              }
    print("Running with params: " + str(params))
    print("Running with ntrees: " + str(ntrees))
    print("Running with features: " + str(features))

    # Train model with local split
    tsize = 0.05
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(train_data, y, test_size=tsize)
    dtrain = xgb.DMatrix(X_train[features].values, np.log(y_train + 1))
    dvalid = xgb.DMatrix(X_test[features].values, np.log(y_test + 1))
    watchlist = [(dvalid, 'eval'), (dtrain, 'train')]
    gbm = xgb.train(params, dtrain, ntrees, evals=watchlist, early_stopping_rounds=100, verbose_eval=True)
    train_probs = gbm.predict(xgb.DMatrix(X_test[features].values))
    indices = train_probs < 0
    train_probs[indices] = 0
    error = rmse_error_(y_test.values, np.exp(train_probs) - 1)
    print(error)

    # Predict and Export
    test_probs = gbm.predict(xgb.DMatrix(test_data[features].values))
    indices = test_probs < 0
    test_probs[indices] = 0
    submission = pd.DataFrame({'Id': test[myid], 'SalePrice': np.exp(test_probs) - 1})
    if not os.path.exists('result/'):
        os.makedirs('result/')
    submission.to_csv("./result/dat-xgb_d%s_eta%s_ntree%s_mcw%s_tsize%s.csv" % (str(depth),str(eta),str(ntrees),str(mcw),str(tsize)) , index=False)
    # Feature importance
#     if plot:
#       outfile = open('xgb.fmap', 'w')
#       i = 0
#       for feat in features:
#           outfile.write('{0}\t{1}\tq\n'.format(i, feat))
#           i = i + 1
#       outfile.close()
#       importance = gbm.get_fscore(fmap='xgb.fmap')
#       importance = sorted(importance.items(), key=operator.itemgetter(1))
#       df = pd.DataFrame(importance, columns=['feature', 'fscore'])
#       df['fscore'] = df['fscore'] / df['fscore'].sum()
#       # Plotitup
#       plt.figure()
#       df.plot()
#       df.plot(kind='barh', x='feature', y='fscore', legend=False, figsize=(25, 15))
#       plt.title('XGBoost Feature Importance')
#       plt.xlabel('relative importance')
#       plt.gcf().savefig('Feature_Importance_xgb_d%s_eta%s_ntree%s_mcw%s_tsize%s.png' % (str(depth),str(eta),str(ntrees),str(mcw),str(tsize)))
```
