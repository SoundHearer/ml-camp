## 3.整体特征-处理结论

### 3-1.整体特征-obj型、数字类别型 处理结论：
- 先做one-hot处理：'MSSubClass','Utilities','Foundation','LotConfig','LandSlope',
                 'BldgType','HouseStyle','Heating','HeatingQC','CentralAir','Electrical'
                 
- 样本比例可能不均衡，后期效果不好再挖掘

### 3-2.整体特征-obj型、数字类别型 处理结论：
- 'OverallQual',曲线关系，看起来还不错，标准化一下，暂不切分
- 'YearBuilt', 'YearRemodAdd'：等距切分
                 
- OverallCond：OverallCond-5 或者 OverallCond-other
