## 数据处理汇总：

## 1.地段特征-结论
- 1.这几个都是类别，直接做一下one-hot-encoding：'MSZoning', ,'Street', 'Neighborhood', 'Condition1', 'Condition2'
- 2.Alley数量太少，缺失值太多的，去掉
- 3.'LotFrontage', 'LotArea'：加了中位数重新分析, 按分布规律分桶，
    - LotFrontage: (0,50) (50,100) (100,150) >150都填充150
    - LotArea: 0,25000 >=25000


## 2. 材料维度-结论
- 2.Roof这两个属性，材料和风格都会有一定影响，RoofMatl有个whsd什么的，可能比较高级，比其他的都影响房价，其他有4种属性固定
- 3.高级材料影响大，到时候考虑加点权重，需要考虑噪声点的处理


- 需要onehot处理这些类别属性：'RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType','ExterQual','ExterCond','LotShape','LandContour'
- MasVnrArea（石板/平方英尺），分桶，多分几个桶==0，!=0

## 4.地下建筑特征处理
- 类别属性统一做one-hot处理：'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1','BsmtFinType2'
- 散点图不够明显，但是也能看出来0和其他的明显有区别
- group按均值（中位数）看看：这次明显了，按分布分桶(数字待校准)
- BsmtFinSF1：0 != 0
- BsmtFinSF2：0 != 0
- 'BsmtUnfSF': (0,1500) >1500
- 'TotalBsmtSF': (0,500) (500,2000) (2000,3300) >3300
    

## 5.地面建筑特征处理
- '1stFlrSF', <400 (400,700) (700,2200) (2200,2500) (2500,3300) 大于3300
- '2ndFlrSF', 0 和 非0
- 'LowQualFinSF', 0 和 非0
- 'GrLivArea' 小于500 (500,2700) (2700,3700) (3700,4700) 大于4700

## 6.户型特征处理
- onehot: KitchenQual Functional
- 离散数值型分类：'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath', 
                   'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual', 'TotRmsAbvGrd','Functional'
  
## 7.配套特征处理：壁炉、车库、露台、走廊、游泳池、围栏
- 类别型变量做one-hot：GarageType GarageFinish GarageQual GarageCond 
- GarageCars虽然数值型，是离散的，也做one-hot处理
- GarageArea（车库面积） 0, (0,200) (200,800) (800,1200), 大于1200
- GarageYrBlt(车库年代) 小于 1940 1940,2000 大于2000
- onehot：Fireplaces,FireplaceQu, PoolArea, PavedDrive
- fence：围栏，有或者没有，分为两类
- 'PoolQC': null太多，直接去掉
- WoodDeckSF   0 和 非0
- OpenPorchSF 0 和 非0
- EnclosedPorch 0 和 非0
- OpenPorchSF 0 和 非0
- ScreenPorch 0 和 非0
