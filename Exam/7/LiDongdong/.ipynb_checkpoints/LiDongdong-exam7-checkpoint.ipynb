#### 1. 机器学习任务中，通常会将给定数据切分为训练集，验证集，测试集，请回答这三类数据集各自的用途。

训练集：用于训练模型时使用，是建立模型时真正使用的数据，生成最终模型的数据    
验证集：用于对训练生成的模型进行验证，验证方法有很多，包括留一法，交叉验证等，这部分数据一般用来调整参数，找到合适参数建立更好模型    
测试集合：测试集合是对最终的结果进行测试，评估最终模型在实际场景中的结果    

#### 2.如何将数据转换成xgboost内置用法的libsvm数据格式，以及该格式数据是如何解读的?

1、xgboost内置用法是将所有的数据格式转化为统一的xgboost所用的格式，该格式是通过xgboost里面的DMatrix()来进行转化的  
2、DMatrix()方法的使用方法为DMatrix(data,label),其中data为训练数据，而label为对应这些数据的标签  

#### 3. 对于自动特征选择，通常有哪三类方法？试写出每类方式的思想，以及在sklearn中的代码实现（面试题）

特征选择可以分为三个部分：  
1、过滤式：过滤式是通过评价特征与结果的相关程度来进行特征选择的，不过这种方法，只是对单个特征与最终结果的评估方法，没有特征与特征之间的相关关系  
基本实现：  
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest
iris = load_iris()
X, y = iris.data, iris.target
X_new = SelectKBest(k=10)
2、包裹式：该种特征选择的方法是通过特征子集的方式来进行的，筛选各种特征子集，评估模型效果，可以通过在全集上不断的减少特征来得到最好的特征子集  
基本实现：  
from sklearn.datasets import load_iris
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rfe = RFE(estimator=rf, n_features_to_select=3)
iris = load_iris()
X, y = iris.data, iris.target
X_rfe = rfe.fit_transform(X,y)     
3、嵌入式：通过模型来评估该特征集合的结果，会对每次得到的特征进行模型评估，效果好的最终作为选择后的特征集合  
基本实现：  
from sklearn.feature_selection import SelectFromModel 
from sklearn.svm import LinearSVC
lsvc = LinearSVC(C=0.02, penalty='l1', dual=False).fit(X,y)
model = SelectFromModel(lsvc, prefit=True)
X_embedd = model.transform(X)

#### 4.网格搜索交叉验证的作用是什么，并简述网格搜索交叉验证是如何运行的？ 

1、网格搜搜交叉验证是通过把可选择的参数进行各种组合，然后选择较优的结果最为参数然后进行交叉验证来进行建模的  
2、运行方式：把能够取得的参数进行组合运算，然后选择组合后的参数挨个加入进行建模和交叉验证，得到结果，每个组合得到一个结果，并选择最优组合得到最终结果

#### 5.工业界上模型整合有三大类方式？试简述每类方式其思想？（面试题）

三大类可以分为：Bagging，Stacking，Boosting  

Bagging：    
1. 给定一个大小为d训练集D  
2. Bagging通过从D中进行可重复的均匀采样 （by sampling from D uniformly and with replacement），得到m个新的训练集Di, 每个Di的大小为n  
3. 将得到的m个新训练集，分别进行回归或者分类，得到m个最终的结果  
4. 由m个结果得到最终结果，对于回归：将m个结果进行平均；对于分类：将m个结果进行投票，票数多的为最终分类值  


Stacking：  
1. 将训练集分成两个不相交的部分  
2. 在第一部分的训练集上训练若干个基本学习器  
3. 在第二部分的训练集上测试得到的基本学习器  
4. 使用步骤3中的预测结果作为输入，将正确的响应（responses）作为输出，训练更高级别的学习器 


Boosting：  
1. 在训练集上，训练一个弱分类器   
2. 添加一个弱分类器进行分类学习   
a. 在这个过程中,由上次的结果进行权值调整, 训练集中数据被重新加权：对错分的样本数据增加权重，对正确分类的样本数据进行降低权重（有些boosting算法会对反复被错分的样本进行降权处理，如BrownBoost和boost by majority）；
b. 在权值调整后的训练集上，进行弱分类器的学习训练  
3. 迭代步骤2   

#### 6.  我们可以将xgboost的中众多参数分类为哪三类？请写出哪些参数可以用什么方式用来控制过拟合？

可以分为：  
1、通用参数：包括输出信息、线程数等  
2、集成参数：包括学习率、正则化参数和树的深度等  
3、任务参数：包括回归类型、评价标准、种子数等  
控制过拟合常用参数：  
正则化项、数的深度、学习率等参数可以控制过拟合

### 1. 使用XGBoost内置方式，导入iris数据完成分类问题（要求以不同参数设置xgboost运行并比对），最后给出实验总结

##### 导入xgboost库和iris数据集

```python
import xgboost as xgb
from sklearn.datasets import load_iris
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

##### 将iris数据集展开为数据和标签部分，并把这部分划分为对应的训练集和测试集

```python
iris = load_iris()
X = iris.data
y = iris.target
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.3,random_state=0)
```

##### 将训练时使用的参数进行指定，并用该参数进行训练

```python
 #参数1
param1 = {'max_depth':2, 'eta':0.15, 'objective':'multi:softmax', 'num_class':3}
dtrain = xgb.DMatrix(data=X_train, label=y_train)
xgb.cv(param, dtrain,20, nfold=3,seed = 1085)
```

##### 从上面的结果可以看出，在20轮训练中，平均的训练错误率从3%逐步降到0.9%，验证集上的错误率基本稳定在3.8%左右，也就是在后面训练中验证集并未取得较好结果，估计是训练的过程中并没有学到更多的具有泛化的性质，需要调整下参数继续进行

##### 重新设置参数

```python
#参数2
param2 = {'max_depth':2, 'eta':0.09, 'objective':'multi:softmax', 'num_class':3}
dtrain = xgb.DMatrix(data=X_train, label=y_train)
xgb.cv(param2, dtrain,30, nfold=3,seed=1090)
```

##### 从上面调整的参数2可以看出，经过30轮训练后，该模型在训练集上的错误率从3.8%逐步下降到0.9%，验证集合上的错误率最终仍旧稳定在了3.8%，此处已经可以考虑在训练集合上出现了过拟合的情况，但是在验证集上仍旧没更好效果，此处考虑增加验证的折数和调节学习率再试试

```python
#参数3
param3 = {'max_depth':2, 'eta':0.2, 'objective':'multi:softmax', 'num_class':3}
dtrain = xgb.DMatrix(data=X_train, label=y_train)
xgb.cv(param3, dtrain,30, nfold=5,seed=1090)
```

##### 从上述结果看出，当前结果已经在训练集上过拟合，验证集上基本达到稳定的程度，此时可以考虑调节参数进行早停止

```python
#参数4
param4 = {'max_depth':2, 'eta':0.2, 'objective':'multi:softmax', 'num_class':3}
dtrain = xgb.DMatrix(data=X_train, label=y_train)
xgb.cv(param4, dtrain,30, nfold=5,early_stopping_rounds=18,seed=1090)
```

#### 确定基本参数和训练方法，开始建立最终的模型

```python
#最终参数
params = {'max_depth':2, 'eta':0.2, 'objective':'multi:softmax', 'num_class':3}
#测试集合
d_test = xgb.DMatrix(data=X_test, label=y_test)
#训练集合
d_train = xgb.DMatrix(data=X_train, label=y_train)
#开始训练
bst = xgb.train(params,d_train)
```

```python
#使用结果进行预测
y_pred = bst.predict(d_test)
```

```python
#结果评价
acc_score = accuracy_score(y_test,y_pred).ravel()
print("The accuracy score is %f" %(acc_score))
```

###### 实验总结：从上面的结果可以看出，经过多次调整参数以后，训练集合上的结果基本上已经达到了最好，甚至训练轮数过度会出现过拟合的情况，此时通过调整学习率可以让学习速度变慢，学的更好，但是容易过拟合，又通过后面调节参数对防止过拟合进行了早停止处理，针对这种情况，在实际场景中可以考虑进行增加训练集合进行尝试，虽然该分类验证集上提升不够明显，但是最终测评结果中准确率还是不错的，达到了98%

### 2. 使用XGBoost的sklearn接口，对KaggleCredit2数据完成信用卡欺诈项目的建模及分析（要求以不同参数设置xgboost运行并比对），最后给出实验总结

- KaggleCredit2数据文件 位于/home/sxy-s3/0.Teacher/Exam/KaggleCredit2.csv.zip，请勿复制或移动该文件

##### 导入需要的包

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV,train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
import zipfile
```

##### 加载数据

```python
with zipfile.ZipFile('/home/sxy-s3/0.Teacher/Exam/KaggleCredit2.csv.zip', 'r') as z:
    f = z.open('KaggleCredit2.csv')
    data = pd.read_csv(f, index_col=0)
data.head()
```

##### 数据处理

```python
y = data['SeriousDlqin2yrs']
X = data.drop('SeriousDlqin2yrs', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

##### 设置参数，进行调参分析

```python
#参数1
param1 = {'max_depth':2, 'eta':0.2, 'objective':'multi:softmax', 'num_class':3}
dtrain = xgb.DMatrix(data=X_train, label=y_train)
xgb.cv(param1, dtrain,30, nfold=5,seed=1090)
```

##### 从上面结果可以看出，当前训练模型设置参数后，训练集和测试集基本错误率都稳定在6%,此时尝试下调整学习率等参数看是否有较好结果

```python
#参数2
param2 = {'max_depth':4, 'eta':0.25, 'objective':'multi:softmax', 'num_class':3}
dtrain = xgb.DMatrix(data=X_train, label=y_train)
xgb.cv(param2, dtrain,30, nfold=8,seed=1090)
```

##### 从上面结果可以看出，把树调深和学习率调节变大以后，结果在验证集和训练集上错误率都出现了下降，效果较之前较好

```python
##利用第二部分调参后的结果进行建模，生成结果
credit_bst = xgb.train(param2, dtrain)
```

```python
#进行预测并进行结果分析
d_test = xgb.DMatrix(data=X_test, label=y_test)
y_pred = credit_bst.predict(d_test)
acc_score = accuracy_score(y_test,y_pred).ravel()
print("The accuracy score is %f" %(acc_score))
```

###### 从上面结果分析可以看出经过进行调参后，使用后面的参数进行了建模，最终结果可以达到94%，总体结果不错
