#### 1. 机器学习任务中，通常会将给定数据切分为训练集，验证集，测试集，请回答这三类数据集各自的用途。

- note:答案不是很理想 ，请参数满分同学

以有监督学习为例来说明，有label的samples，用来训练模型的samples的集合称为训练集，需要预测的无label的samples称作测试集。在训练得到模型之后，并不能知道在测试集上的性能如何，而且有的模型有许多备选参数，在不同参数下，训练出来的模型也不同，最终在测试集山的效果也不同，所以将有label的数据分为训练集合验证集，训练集负责训练模型，在验证集上验证该模型的好坏。切分验证集的方法有留出法（包括留一法）和交叉验证法。

#### 3. 对于自动特征选择，通常有哪三类方法？试写出每类方式的思想，以及在sklearn中的代码实现（面试题）

- note:缺 代码

特征选择通常有三种形式，过滤型、包裹型、嵌入型。    
过滤型的选择方式，将特征按照重要度或者和标签的相关性进行排序，取前k个重要的特征或者重要度大于某个值的特征，这些特征构成训练模型的训练集，整个筛选的过程和选用哪种模型无关。    
包裹型的选择方式和准备使用的模型有关，特征的选择取决于该特征的加入和删除是否影响模型的结果。一般有前向搜索和后向搜索，前向搜索的意思是给定特征子集，将某个特征加入该特征子集，对模型进行训练，如果特征效果变好，则保留该特征，否则删除。后向搜索是对给定特征子集，和已经训练好的模型，减去某个特征，查看对模型的效果是否有影响，效果差异性很大，则说明该特征相关性或者比重较高，需要保留。      
嵌入型的选择方式是将特征选择与模型的训练融合为一体，在训练模型的同时，对特征进行了选择。

#### 4.网格搜索交叉验证的作用是什么，并简述网格搜索交叉验证是如何运行的？ 

模型训练时，需要手动调节一些参数，从这些参数中找出模型最优。网格搜索程序自动完成这个事情，即在给定的几组参数中，分别训练一次，找出效果最好的模型参数。在验证模型效果时，训练集和验证集切分的不同，训练出来的模型及效果也不同，而且只由一次切分并不足以说明该模型的效果好，所以引入交叉验证。k折交叉验证是将有label的数据分为k份，取其中1份作为验证集，k-1份作为训练集，训练模型，在验证集上验证。这样的操作进行k次，取k次训练结果的平均或者投票，得出最终效果评分。然后由网格搜索找出效果最佳的那一组参数。

#### 5.工业界上模型整合有三大类方式？试简述每类方式其思想？（面试题）

Bagging、Voting、Random forest       
Stacking和blending     
Boosting   
以Bagging为代表的思想是投票或者加权，每个基学习器会对数据进行学习得到一个分值或者标签，将每个学习器的结果进行加权或者投票得到最终的结果。     
Stacking与Blending的思想是在上一层模型的基础上做一个融合，blending是线性加权。  在上一层几个模型的基础上，使用一个模型来预测真实值。
Boosting的思想是优化预测错误样本的概率或者个数上，即下一个模型是基于上一个模型预测错误的样本，目的是将预测错误样本概率或者个数减少。    

#### 6.  我们可以将xgboost的中众多参数分类为哪三类？请写出哪些参数可以用什么方式用来控制过拟合？

模型参数分为3类，有通用参数、集成（增强）参数和任务参数。    
集成参数中的几个可以控制过拟合，包括gamma、max_depth、min_child_weight、subsample、colsample_bytree。    
gamma用来调节作进一步分割时，最少的损失减少值，如果小于该值，则暂停往下生长，如果该值设置的很小，那么就会将一些离群点或者噪声点的特征学习进来，适当的配置这个值，可以控制过拟合。    
max_depth用来设置树的最大生长深度，即树生长到深度n之后，不再继续生长，适当的配置这个可以控制过拟合。    
min_child_weight用来调节进一步生成节点的时候，如果观测的样本权重小于这个数值时，则停止生长，防止分的太细，导致将离群点和噪声点的特征学习进来，或者将训练集的特征学习太好，而不是测试集的特征属性，导致过拟合。适当的配置这个参数可以控制过拟合。   
subsample设置用于生长的观测子样本的比例，从当前样本中抽取多少样本进行生长，这样可以引入样本的随机性，多样性，可以有效抑制过拟合。

### 2. 使用XGBoost的sklearn接口，对KaggleCredit2数据完成信用卡欺诈项目的建模及分析（要求以不同参数设置xgboost运行并比对），最后给出实验总结

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import KFold,train_test_split,GridSearchCV
from sklearn.metrics import confusion_matrix,mean_squared_error
```

```python
credit_df = pd.read_csv('./KaggleCredit2.csv')
credit_df.head()
```

```python
X = credit_df.drop(['SeriousDlqin2yrs'],axis=1)
y = credit_df['SeriousDlqin2yrs']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
# 从类别数量可以看出0，1的分布十分不平衡
credit_df['SeriousDlqin2yrs'].value_counts()
```

使用一个base的Xgboost的参数值，可以看到最好结果是在0.0650左右,最后的score是0.9348487084870849

```python
# 先设置一个初始的Xgboost的参数值，跑一下试试
xgb_model_base = xgb.XGBClassifier(max_depth=4,learning_rate=0.1,\
                                  n_estimators=200,objective='binary:logistic',\
                                  n_jobs=2,gamma=0.01,min_child_weight=1,\
                                  subsample=0.8,colsample_bytree=0.9,\
                                  scale_pos_weight=1)
xgb_model_base.fit(X_train,y_train,eval_set=[(X_test,y_test)],eval_metric='error')
xgb_model_base.score(X_test,y_test)
```
