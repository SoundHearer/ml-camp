#### 1. 机器学习任务中，通常会将给定数据切分为训练集，验证集，测试集，请回答这三类数据集各自的用途。

- Training set（训练集）: 
    - A set of examples used for learning, which is to fit the parameters [i.e., weights] of the classifier. 
    - 一个样本的数据集，用来（给模型进行）学习的示例数据，用来适配模型（分类器还有回归器）的参数（比如权重）。
    
- Validation set（验证集）: 
    - A set of examples used to tune the parameters [i.e., architecture, not weights] of a classifier, for example to choose the number of hidden units in a neural network.
    - 一个样本的数据集，用来调试模型的（分类器或者回归器等）参数，比如选择一个神经网络的隐藏单元的个数。
    
    
- Test set（测试集）: 
    - A set of examples used only to assess the performance [generalization] of a fully specified classifier.
    - 一个样本的数据集，用来评估一个完全特征化的模型（如分类器）的表现（通用性）。

- 1.Filter：过滤法，按照发散性或者相关性对各个特征进行评分，设定阈值或者待选择阈值的个数，选择特征。
    - 如：卡方检验（SelectKBest），经典的卡方检验是检验定性自变量对定性因变量的相关性。假设自变量有N种取值，因变量有M种取值，考虑自变量等于i且因变量等于j的样本频数的观察值与期望的差距，构建统计量

```python
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest

iris = load_iris()
X = iris.data
y = iris.target
print(X.shape)
k_best = SelectKBest(k=2)
k_best
```

- 2.wrapper：包裹式（包装法），根据目标函数（通常是预测效果评分），每次选择若干特征，或者排除若干特征。
    - 递归特征消除法: 递归消除特征法使用一个基模型来进行多轮训练，每轮训练后，消除若干权值系数的特征，再基于新的特征集进行下一轮训练。使用feature_selection库的RFE类来选择特征的代码如下
  

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

rfe = RFE(estimator=RandomForestClassifier(), n_features_to_select=2)
X_rfe = rfe.fit_transform(X,y)
X_rfe.shape
```

#### 4.网格搜索交叉验证的作用是什么，并简述网格搜索交叉验证是如何运行的？ 

网格搜索交叉验证，目标是找到最优参数组合，提供多组备选参数，用交叉验证的方式反复验证参数组合，选出综合效果最好（如评分等）的一组参数。

#### 5.工业界上模型整合有三大类方式？试简述每类方式其思想？（面试题）

- Voting，投票，每个模型都有预测，多个模型投票得出结果，得票最多的即为结果
- Bagging，Bagging算法可以并行处理，Bagging就是采用有放回的方式进行抽样，用抽样的样本建立子模型,对子模型进行训练，这个过程重复多次，最后进行融合。
- Boosting，而Boosting的思想是一种迭代的方法，每一次训练的时候都更加关心分类错误的样例，给这些分类错误的样例增加更大的权重，下一次迭代的目标就是能够更容易辨别出上一轮分类错误的样例。最终将这些弱分类器进行加权相加。

- 2.集成(增强)参数/booster parameters
- 3.任务参数/task parameters

* 控制过拟合：减小eta、增大gamma、降低tree depth、subsample、调整正则系数（lambda、alpha）

### 加载数据：可添加路径参数数据类型，更通用

```python
def load_data():
    print('-------------------start load data...')
    import zipfile
    with zipfile.ZipFile('./KaggleCredit2.csv.zip', 'r') as z:
        file = z.open('KaggleCredit2.csv')
        data = pd.read_csv(file, index_col=0)
    print('-------------------load data success.')
    return data
```

### 准备数据，准备X, y, train, test数据集以及xgb.DMatrix

```python
def prepare_data(data):
    print('-------------------start prepare data...')
    
    from sklearn.model_selection import train_test_split
    
    data.dropna(inplace=True)
    data.shapey = data['SeriousDlqin2yrs']
    
    

    X = data.drop('SeriousDlqin2yrs', axis=1)
    y = data['SeriousDlqin2yrs']
    X = data.drop('SeriousDlqin2yrs', axis=1)
    print('X.shape, y.shape:', X.shape,y.shape)
    print('----------------------------------------')
    print('y.value_counts():', y.value_counts())
    print('----------------------------------------')
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

    print('X_train.shape, X_test.shape, y_train.shape, y_test.shape: ' ,X_train.shape, X_test.shape, y_train.shape, y_test.shape)

    d_train = xgb.DMatrix(data=X_train , label=y_train)
    d_test = xgb.DMatrix(data=X_test, label=y_test)
    print('-------------------data done!')
    return d_train, d_test, X_train, X_test, y_train, y_test 
```
