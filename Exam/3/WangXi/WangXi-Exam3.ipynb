4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
df[['G1','G2','G3']].groupby(df['sex']).aggregate(np.mean)
```

5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

***
Done!   
绘图比其它同学用心了。

```python
dfAge = df[['G1','G2','G3']].groupby(df['age']).aggregate(np.mean)
dfAge
```

```python
ages = list(dfAge.index)
g1 = pd.Series((dfAge['G1']))
g2 = pd.Series((dfAge['G2']))
g3 = pd.Series((dfAge['G3'])) 

w = 0.3
plt.figure(figsize=(9,6))
plt.bar([i-w for i in ages], g1, width=0.3, label='G1',facecolor = 'm', edgecolor = 'white')
plt.bar([i for i in ages], g2, width=0.3, label='G2',facecolor = 'yellowgreen')
plt.bar([i+ w for i in ages], g3,width=0.3, label='G3',facecolor = 'lightskyblue')
plt.xticks(fontsize =12)  
plt.yticks(fontsize =12)  
plt.xlabel('Ages',fontsize =16)    
plt.ylabel('Average Scores',fontsize =16)    
plt.title('Average Scores by Group of Age',fontsize =16)  
plt.legend(loc='upper right',fontsize =16)
plt.tight_layout()  
plt.show()
```

6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
dfSchool = df[['G1','G2','G3']].groupby(df['school']).aggregate(np.mean)
dfSubject = dfSchool.T
dfSubject
```

```python
dfSchool = df[['G1','G2','G3']].groupby(df['school']).aggregate(np.mean)
dfSubject = dfSchool.T
n = len(dfSubject.index)
index = range(n)  
g1 = pd.Series((dfSubject['GP']))
g2 = pd.Series((dfSubject['MS']))

plt.figure(figsize=(9,6))
plt.bar([i-0.2 for i in index], g1, width=0.4, label='GP',facecolor = 'lightskyblue', edgecolor = 'white')
plt.bar([i+0.2 for i in index], g2, width=0.4, label='MS',facecolor = 'yellowgreen')
plt.xticks(index,('G1','G2','G3'),fontsize =15)   
plt.yticks(fontsize =15) 
plt.xlabel('Subjects',fontsize =16)    
plt.ylabel('Average Scores',fontsize =16)    
plt.title('Average Scores by Group of Shcool',fontsize =16)  
plt.legend(loc='upper right',fontsize =13)
plt.tight_layout()  
plt.show()
```

7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(40分)

```python
df['sum'] = df['G1']+df['G2']+df['G3']

def topTen(df, n=10, column='sum'):
#     return df.sort_values(by=column, ascending=False)[:n]
      return df.nlargest(n,column)

groups = df.groupby(df['school']).apply(topTen)

groups.to_csv("ranking.csv",index=False,sep=';')
groups['ranking'] = list(range(1,11))+list(range(1,11))
groups.set_index(['school','ranking'])
```
