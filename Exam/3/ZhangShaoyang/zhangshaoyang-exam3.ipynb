### Pandas实战题(共七个步骤，每个步骤要求及分数均已标明。总分100分)

1.请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（7.5分）   
请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（7.5分）

```python
import pandas as pd
import numpy as np
df_info = pd.read_csv('student-info.csv',sep = ';')
df_info.head(5)
```

```python
df_score = pd.read_csv('student-score.csv',sep = ';')
df_score.head(5)
```

2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (10分）

```python
df = pd.merge(df_info, df_score, on = 'ID')
df.head()
```

3.把"df"所有的列展示出来(5分)

```python
df.columns
```

```python
for i in df.columns:
    print(i)
```

4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
df.groupby('sex').agg({'G1':np.mean, 'G2':np.mean,'G3':np.mean})
```

5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
%matplotlib inline
df_age_group = df.groupby('age').agg({'G1':np.mean, 'G2':np.mean,'G3':np.mean})
df_age_group.plot(kind = 'bar',figsize = (20,5),rot = 'horizontal')
df_age_group
```

6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
df_school_group = df.groupby('school').agg({'G1':np.mean, 'G2':np.mean,'G3':np.mean}).T.reindex(['G1','G2','G3']) 
df_school_group.plot(kind = 'bar',rot = 'horizontal')
df_school_group
```

7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(40分)

```python
df['G_score'] = df['G1']+df['G2']+df['G3']
df_top10 = df.groupby('school').apply(lambda x : x.nlargest(10,'G_score'))
```
