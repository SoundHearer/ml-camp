### Pandas实战题(共七个步骤，每个步骤要求及分数均已标明。总分100分)

1.请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（7.5分）   
请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（7.5分）

```python
import numpy as np
import pandas as pd
df_info=pd.read_csv('../data/student-info.csv',sep=';')
df_info.head(5)
df_score=pd.read_csv('../data/student-score.csv',sep=';')
df_score.head(5)
```

2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (10分）

***
Not Done!   
虽然这样操作可以，但初学阶段，请指定是left,right，是inner,还是outer,

```python
df=pd.merge(df_info,df_score)
df.head()
```

3.把"df"所有的列展示出来(5分)

***
Not Done!   
这种展示出来，但我们可以直接用.columns就可以的呀，为啥还要循环呢

```python
for x in df.columns:
    print(x)
```

4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
%matplotlib inline
df1=df.groupby('sex')
df1.aggregate({'G1':np.mean,'G2':np.mean,'G3':np.mean})
```

5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
df2=df.groupby('age')
df2.aggregate({'G1':np.mean,'G2':np.mean,'G3':np.mean}).plot(kind='bar')
```

6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

***
Not Done!   
会用subplot很好，只是这题的可视化并没有很好的完成题目上中的要求。

```python
df3=df.groupby('school')
df3.aggregate({'G1':np.mean,'G2':np.mean,'G3':np.mean}).plot(kind='bar',subplots=True)
```
