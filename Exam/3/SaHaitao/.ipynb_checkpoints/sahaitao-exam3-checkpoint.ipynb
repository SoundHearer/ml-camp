请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（7.5分）

```python
import pandas as pd
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = "svg"
```

```python
df_info = pd.read_csv("../data/student-info.csv", sep=";", index_col=0)
df_info.head()
```

```python
df_score = pd.read_csv("../data/student-score.csv", sep=";", index_col=0)
df_score.head()
```

2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (10分）

```python
df = df_info.join(df_score)
```

3.把"df"所有的列展示出来(5分)

4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
group_by_sex = df.groupby("sex")
mean_value = group_by_sex.agg({key:np.mean for key in ("G1","G2","G3")})
mean_value
```

5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
group_by_age = df.groupby("age")
group_by_age.agg({key:np.mean for key in ("G1","G2","G3")}).plot(kind="bar")
```

6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
group_by_school = df.groupby("school")
aver_by_school = group_by_school.agg({key:np.mean for key in ("G1","G2","G3")})
print(aver_by_school)
aver_by_school.T.plot(kind="bar", rot=90)
```
