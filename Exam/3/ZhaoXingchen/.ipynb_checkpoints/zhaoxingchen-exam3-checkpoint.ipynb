### Pandas实战题(共七个步骤，每个步骤要求及分数均已标明。总分100分)

- 所需的两个数据文件在本目录下，请复制至您的作业目录中。

1.请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（7.5分）   
请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（7.5分）

```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
# 读student-info.csv
df_info = pd.read_csv("student-info.csv",sep = ';')
df_info = pd.DataFrame(df_info)
df_info.head(5)
```

```python
# 读student-score.csv
df_score = pd.read_csv("student-score.csv",sep = ';')
df_score = pd.DataFrame(df_score)
df_score.head(5)
```

2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (10分）

```python
df = pd.concat([df_info,df_score],axis = 1)
df.head(20)
```

3.把"df"所有的列展示出来(5分)

```python
list(df.columns)
```

4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
df.groupby('sex')['sex','G1','G2','G3'].head()
# df.groupby(['sex'])['G1','G2','G3'].head()
# df.dtypes
df.groupby('sex')['sex','G1','G2','G3'].agg({"Average":np.mean})
```

5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
age_score = df.groupby('age')['G1','G2','G3']
age_score.head()
age_score.agg({"Average":np.mean})
```

```python
# 柱状图
age_score = age_score.agg({"Average":np.mean})
# figsize = (30,20)
age_score.plot.bar(grid = True,fontsize =15)
```

6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
# school_score = df.groupby('school')['G1','G2','G3']
# school_score.head()
# school_score.agg({"Average":np.mean})

school_score = df.groupby('school').mean()
school_score = school_score[['G1','G2','G3']]
school_score.T.plot.bar()
```

7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(40分)

```python
# top10_score = df.groupby(['school'])['G1','G2','G3']  .groupby('school')
df_g = df.loc[:,['school','G1','G2','G3']].T
df_g = pd.DataFrame(df_g.drop('school').apply(sum))
df_g.columns = ['sum']
df_g.head(20)
```

```python
top10_score =pd.DataFrame(df_g["sum"].nlargest(10))
top10_score.index
```

```python
df.head()
top10_info = pd.DataFrame(df,index=top10_score.index)
top10_info.drop('ID',axis = 1)
```
