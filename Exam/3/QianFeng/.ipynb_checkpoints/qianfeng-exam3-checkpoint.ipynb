### Pandas实战题(共七个步骤，每个步骤要求及分数均已标明。总分100分)

1.请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（7.5分）   
请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（7.5分）

```python
import pandas as pd

df_info = pd.read_csv('../data/student-info.csv', sep =';')
df_info.head(5)
# df_info  # 395 rows × 31 columns
```

```python
df_score = pd.read_csv('../data/student-score.csv', sep =';')
df_score.head(5)
# df_score # 395 rows × 4 columns
```

2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (10分）

```python
df = pd.merge(df_info, df_score, on='ID', how='outer')
df  # 395 rows × 34 columns
```

3.把"df"所有的列展示出来(5分)

```python
# dir(df)
df.columns
```

4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
import numpy as np

df.groupby('sex').aggregate({'G1': np.mean, 'G2': np.mean, 'G3': np.mean})
```

5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
import matplotlib.pyplot as plt
%matplotlib inline

df.groupby('age').aggregate({'G1': np.mean, 'G2': np.mean, 'G3': np.mean}).plot(kind='bar')
```

6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
result = df.groupby('school').aggregate({'G1': np.mean, 'G2': np.mean, 'G3': np.mean}).T
result.plot(kind='bar')
```

7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(40分)

```python
def f(group):
    return group.nlargest(10,'sum')

df['sum'] = df['G1'] + df['G2'] + df['G3']
df.groupby('school').apply(f)   # 存在并列第十名, 跑的结果和本地竟然不一样
```
