### Pandas实战题(共七个步骤，每个步骤要求及分数均已标明。总分100分)

1.请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（7.5分）   
请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（7.5分）

```python
import pandas as pd
```

```python
df_info = pd.read_csv("../data/student-info.csv",sep=';',index_col=0)
df_score = pd.read_csv("../data/student-score.csv",sep=';',index_col=0)
```

```python
df_info.head()
```

```python
df_score.head()
```

2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (10分）

```python
df = pd.concat([df_info,df_score],axis=1)
df.head()
```

3.把"df"所有的列展示出来(5分)

```python
df.columns
```

4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
import numpy as np
df_sex_score = pd.concat([df['sex'],df['G1'],df['G2'],df['G3']],axis=1)
sex_ave = df_sex_score.groupby('sex').agg([np.mean])
sex_ave
```

5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
df_age_score = pd.concat([df['age'],df['G1'],df['G2'],df['G3']],axis=1)
age_ave = df_age_score.groupby('age').agg([np.mean])
age_ave
```

```python
%matplotlib inline
age_ave.plot(kind='bar')
```

6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
df_school_score = pd.concat([df['school'],df['G1'],df['G2'],df['G3']],axis=1)
school_ave = df_school_score.groupby('school').agg([np.mean])
school_ave
```

```python
school_ave.T.plot(kind='bar')
```

7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(40分)

```python
df['zongfen'] = df['G1']+df['G2']+df['G3']
df1 = df[df.school=='GP']
df1.sort_values('zongfen',ascending=False).head(10)
```

```python
df2 = df[df.school=='MS']
df2.sort_values('zongfen',ascending=False).head(10)
```
