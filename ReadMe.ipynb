# 课程总结


## 第一周（Python基础）
#### 背景知识要求：
- 基本编程知识

#### 学习目标:
- 云平台Notebook的熟练运行Python
- Python的基础语法
- 循环控制，函数及面向对向，
- Python基础语法及进阶知识
- Python函数，类，对象的操作
- 内置及第三方库的应用


## 第二周（数据抓取）
#### 背景知识要求：
- HTML基础知识
- 数据库基础知识

#### 学习目标:
- 使用requests库爬取数据
- 使用scrapy框架搭建爬虫工程
- 部分数据库操作


## 第三周（数据处理）

#### 背景知识要求：
- Python基础操作
- 有数据可视化的基本概念

#### 学习目标:
- 熟练掌握Pandas库的对于数据的各类操作

## 第四周（数据可视化及拟合）

#### 背景知识要求：
- Python及Pandas部分要求掌握
- 了解不同数据类型
- 对数据可视化的意义有了解
- 对数据可视化的不同实现方案有大致了解

#### 学习目标：
- matplotlib的使用
- 基于matplotlib库的高级绘图seaborn的使用


## 第五周（大数据部分）

#### 背景知识要求：
- linux基本操作
- 了解大数据应用背景

#### 学习目标
- 了解海量数据的处理方式
- Hadoop集群完成Map-reduce任务
- Spark进行RDD操作


## 第六周（机器学习部分）

#### 背景知识要求：
- 该阶段主要学习对之前Python基础及数据处理部分有要求

#### 学习目标
- 机器学习基础概念
- 机器学习的完整流程

## 第七周（模型融合）

#### 背景知识要求：
- 该阶段主要学习对之前python,数据处理，机器学习，算法等进行综合应用

#### 学习目标
- 模型集成的基本概念，操作
